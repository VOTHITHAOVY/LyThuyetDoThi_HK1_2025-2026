```python
# B√ÄI T·∫¨P ·ª®NG D·ª§NG - CH·∫†Y TR√äN GOOGLE COLAB

print("=== B√ÄI T·∫¨P ·ª®NG D·ª§NG ===\n")

# B√†i 1: B·ªô ba
print("üéØ B√ÄI 1: B·ªò BA")
def bai1_demo():
    test_cases = [
        "DAIHOCKHOAHOCTUNHIENTHANHPHOHOCHIMINH",
        "AAAAAD",
        "ABCABCDABC"
    ]

    for s in test_cases:
        if len(s) < 3:
            print(f"'{s}' -> 0")
            continue

        freq = {}
        max_count = 0

        for i in range(len(s) - 2):
            triple = s[i:i+3]
            freq[triple] = freq.get(triple, 0) + 1
            max_count = max(max_count, freq[triple])

        most_common = [k for k, v in freq.items() if v == max_count]
        print(f"'{s}' -> {max_count} (b·ªô ba: {most_common})")

bai1_demo()

# B√†i 2: D√£y ho√°n v·ªã
print("\nüéØ B√ÄI 2: D√ÉY HO√ÅN V·ªä")
def bai2_demo():
    test_cases = [
        ([4], [1, 3, 2, 4]),
        ([4], [1, 2, 2, 4]),
        ([3], [1, 2, 5]),
        ([5], [1, 2, 3, 4, 5])
    ]

    for (N, arr) in test_cases:
        visited = [False] * (N[0] + 1)
        is_permutation = True

        for num in arr:
            if num < 1 or num > N[0]:
                is_permutation = False
                break
            if visited[num]:
                is_permutation = False
                break
            visited[num] = True

        for i in range(1, N[0] + 1):
            if not visited[i]:
                is_permutation = False
                break

        print(f"N={N[0]}, d√£y={arr} -> {'YES' if is_permutation else 'NO'}")

bai2_demo()

# B√†i 3: Ph√¢n t√≠ch N!
print("\nüéØ B√ÄI 3: PH√ÇN T√çCH N!")
def sieve_eratosthenes(n):
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            for j in range(i*i, n+1, i):
                is_prime[j] = False

    return [i for i in range(2, n+1) if is_prime[i]]

def factorial_factorization(n):
    primes = sieve_eratosthenes(n)
    exponents = []

    for p in primes:
        exp = 0
        power = p
        while power <= n:
            exp += n // power
            power *= p
        exponents.append(exp)

    return exponents

def bai3_demo():
    test_cases = [10, 5, 6, 8]

    for n in test_cases:
        result = factorial_factorization(n)
        print(f"{n}! = ", end="")

        primes = sieve_eratosthenes(n)
        factors = [f"{primes[i]}^{result[i]}" for i in range(len(primes)) if result[i] > 0]
        print(" * ".join(factors))
        print(f"Bi·ªÉu di·ªÖn: {' '.join(map(str, result))}")

bai3_demo()

# B√†i 4: S·ªë ch·ªØ s·ªë 0
print("\nüéØ B√ÄI 4: S·ªê CH·ªÆ S·ªê 0 T·∫¨N C√ôNG")
def count_trailing_zeros(n):
    count = 0
    divisor = 5

    while divisor <= n:
        count += n // divisor
        divisor *= 5

    return count

def bai4_demo():
    test_cases = [5, 10, 25, 100, 1, 15]

    for n in test_cases:
        zeros = count_trailing_zeros(n)
        print(f"{n}! c√≥ {zeros} ch·ªØ s·ªë 0 t·∫≠n c√πng")

bai4_demo()

# B√†i 5: Ph·∫ßn t·ª≠ chung
print("\nüéØ B√ÄI 5: PH·∫¶N T·ª¨ CHUNG")
def bai5_demo():
    test_cases = [
        (5, [1, 2, 5, 4, 3], 4, [5, 6, 1, 4], 4, [5, 3, 5, 1]),
        (3, [1, 2, 3], 3, [2, 3, 4], 3, [3, 4, 5]),
        (4, [10, 20, 30, 40], 4, [20, 30, 40, 50], 4, [30, 40, 50, 60])
    ]

    for nx, x, ny, y, nz, z in test_cases:
        max_val = 10000
        count_x = [False] * (max_val + 1)
        count_y = [False] * (max_val + 1)
        count_z = [False] * (max_val + 1)

        for num in x:
            if num <= max_val:
                count_x[num] = True

        for num in y:
            if num <= max_val:
                count_y[num] = True

        for num in z:
            if num <= max_val:
                count_z[num] = True

        common = []
        for i in range(1, max_val + 1):
            if count_x[i] and count_y[i] and count_z[i]:
                common.append(i)

        print(f"D√£y x: {x}")
        print(f"D√£y y: {y}")
        print(f"D√£y z: {z}")
        print(f"‚Üí Ph·∫ßn t·ª≠ chung: {common} (s·ªë l∆∞·ª£ng: {len(common)})\n")

bai5_demo()

print("=== K·∫æT TH√öC ===")
```
