```python
# -*- coding: utf-8 -*-
"""CAY_KHUNG_NHO_NHAT_FULL.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1abc123
"""

# C√ÄI ƒê·∫∂T TH∆Ø VI·ªÜN
!pip install networkx matplotlib

import heapq
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from collections import deque, defaultdict

print("=== TH∆Ø VI·ªÜN ƒê√É S·∫¥N S√ÄNG ===\n")

# ============================================================================
# B√ÄI 1: C√ÇY KHUNG DFS
# ============================================================================

print("="*60)
print("B√ÄI 1: C√ÇY KHUNG DFS")
print("="*60)

def dfs_spanning_tree(n, edges):
    """T√¨m c√¢y khung b·∫±ng DFS"""
    # X√¢y d·ª±ng danh s√°ch k·ªÅ
    adj = [[] for _ in range(n + 1)]
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    visited = [False] * (n + 1)
    spanning_tree = []
    stack = [1]
    visited[1] = True

    print("üîç Qu√° tr√¨nh DFS:")
    step = 1

    while stack:
        u = stack.pop()
        print(f"B∆∞·ªõc {step}: X√©t ƒë·ªânh {u}")

        for v in adj[u]:
            if not visited[v]:
                visited[v] = True
                spanning_tree.append((u, v))
                stack.append(v)
                print(f"  ‚Üí Th√™m c·∫°nh ({u}, {v})")

        step += 1

    return spanning_tree

# Test B√†i 1
print("\nüìù TEST B√ÄI 1 - C√ÇY KHUNG DFS")
n1, m1 = 6, 6
edges1 = [(1, 2), (1, 3), (1, 6), (2, 4), (2, 5), (5, 6)]

print(f"ƒê·ªì th·ªã: {n1} ƒë·ªânh, {m1} c·∫°nh")
print(f"C√°c c·∫°nh: {edges1}")

dfs_tree = dfs_spanning_tree(n1, edges1)
print(f"\n‚úÖ K·∫æT QU·∫¢:")
print(f"S·ªë c·∫°nh trong c√¢y khung: {len(dfs_tree)}")
print(f"C√°c c·∫°nh: {dfs_tree}")

# V·∫Ω ƒë·ªì th·ªã B√†i 1
G1 = nx.Graph()
G1.add_edges_from(edges1)
pos1 = nx.spring_layout(G1, seed=42)

plt.figure(figsize=(10, 6))
nx.draw(G1, pos1, with_labels=True, node_color='lightblue',
        node_size=500, font_size=10, font_weight='bold')
nx.draw_networkx_edges(G1, pos1, edgelist=dfs_tree,
                      edge_color='red', width=2)
plt.title("B√ÄI 1: C√ÇY KHUNG DFS", fontsize=14, fontweight='bold')
plt.show()

# ============================================================================
# B√ÄI 2: C√ÇY KHUNG BFS
# ============================================================================

print("\n" + "="*60)
print("B√ÄI 2: C√ÇY KHUNG BFS")
print("="*60)

def bfs_spanning_tree(n, edges):
    """T√¨m c√¢y khung b·∫±ng BFS"""
    adj = [[] for _ in range(n + 1)]
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    visited = [False] * (n + 1)
    spanning_tree = []
    queue = deque([1])
    visited[1] = True

    print("üîç Qu√° tr√¨nh BFS:")
    step = 1

    while queue:
        u = queue.popleft()
        print(f"B∆∞·ªõc {step}: X√©t ƒë·ªânh {u}")

        for v in adj[u]:
            if not visited[v]:
                visited[v] = True
                spanning_tree.append((u, v))
                queue.append(v)
                print(f"  ‚Üí Th√™m c·∫°nh ({u}, {v})")

        step += 1

    return spanning_tree

# Test B√†i 2
print("\nüìù TEST B√ÄI 2 - C√ÇY KHUNG BFS")
bfs_tree = bfs_spanning_tree(n1, edges1)

print(f"\n‚úÖ K·∫æT QU·∫¢:")
print(f"S·ªë c·∫°nh trong c√¢y khung: {len(bfs_tree)}")
print(f"C√°c c·∫°nh: {bfs_tree}")

# V·∫Ω ƒë·ªì th·ªã B√†i 2
plt.figure(figsize=(10, 6))
nx.draw(G1, pos1, with_labels=True, node_color='lightblue',
        node_size=500, font_size=10, font_weight='bold')
nx.draw_networkx_edges(G1, pos1, edgelist=bfs_tree,
                      edge_color='green', width=2)
plt.title("B√ÄI 2: C√ÇY KHUNG BFS", fontsize=14, fontweight='bold')
plt.show()

# ============================================================================
# B√ÄI 3: THU·∫¨T TO√ÅN KRUSKAL
# ============================================================================

print("\n" + "="*60)
print("B√ÄI 3: THU·∫¨T TO√ÅN KRUSKAL")
print("="*60)

class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n + 1))
        self.rank = [0] * (n + 1)

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)

        if rootX == rootY:
            return False

        if self.rank[rootX] < self.rank[rootY]:
            self.parent[rootX] = rootY
        elif self.rank[rootX] > self.rank[rootY]:
            self.parent[rootY] = rootX
        else:
            self.parent[rootY] = rootX
            self.rank[rootX] += 1

        return True

def kruskal(n, edges):
    """T√¨m c√¢y khung nh·ªè nh·∫•t b·∫±ng Kruskal"""
    print("üìä Danh s√°ch c·∫°nh ban ƒë·∫ßu:")
    for i, (u, v, w) in enumerate(edges):
        print(f"  C·∫°nh {i}: ({u}, {v}) - tr·ªçng s·ªë {w}")

    # S·∫Øp x·∫øp c·∫°nh theo tr·ªçng s·ªë
    sorted_edges = sorted(edges, key=lambda x: x[2])
    print(f"\nüî¢ C·∫°nh sau khi s·∫Øp x·∫øp:")
    for i, (u, v, w) in enumerate(sorted_edges):
        print(f"  {i+1}. ({u}, {v}) - {w}")

    uf = UnionFind(n)
    mst = []
    total_weight = 0

    print(f"\nüîÑ Qu√° tr√¨nh x√¢y d·ª±ng MST:")
    for i, (u, v, w) in enumerate(sorted_edges):
        if uf.union(u, v):
            mst.append((u, v, w))
            total_weight += w
            print(f"  ‚úÖ Th√™m ({u}, {v}, {w}) - T·ªïng: {total_weight}")
        else:
            print(f"  ‚ùå B·ªè qua ({u}, {v}, {w}) - T·∫°o chu tr√¨nh")

    return mst, total_weight

# Test B√†i 3
print("\nüìù TEST B√ÄI 3 - KRUSKAL")
n3 = 5
edges3 = [(1, 2, 1), (1, 3, 2), (2, 3, 3),
          (1, 4, 9), (3, 4, 7), (4, 5, 2)]

kruskal_mst, kruskal_weight = kruskal(n3, edges3)

print(f"\n‚úÖ K·∫æT QU·∫¢:")
print(f"S·ªë c·∫°nh: {len(kruskal_mst)}, T·ªïng tr·ªçng s·ªë: {kruskal_weight}")
print(f"C√°c c·∫°nh: {kruskal_mst}")

# V·∫Ω ƒë·ªì th·ªã B√†i 3
G3 = nx.Graph()
for u, v, w in edges3:
    G3.add_edge(u, v, weight=w)

pos3 = nx.spring_layout(G3, seed=42)
plt.figure(figsize=(10, 6))
nx.draw(G3, pos3, with_labels=True, node_color='lightblue',
        node_size=500, font_size=10, font_weight='bold')

# V·∫Ω t·∫•t c·∫£ c·∫°nh
nx.draw_networkx_edges(G3, pos3, alpha=0.3)

# V·∫Ω MST
mst_edges = [(u, v) for u, v, w in kruskal_mst]
nx.draw_networkx_edges(G3, pos3, edgelist=mst_edges,
                      edge_color='red', width=2)

# V·∫Ω tr·ªçng s·ªë
edge_labels = {(u, v): w for u, v, w in edges3}
nx.draw_networkx_edge_labels(G3, pos3, edge_labels=edge_labels)

plt.title("B√ÄI 3: KRUSKAL - C√ÇY KHUNG NH·ªé NH·∫§T", fontsize=14, fontweight='bold')
plt.show()

# ============================================================================
# B√ÄI 4: THU·∫¨T TO√ÅN PRIM
# ============================================================================

print("\n" + "="*60)
print("B√ÄI 4: THU·∫¨T TO√ÅN PRIM")
print("="*60)

def prim(n, edges):
    """T√¨m c√¢y khung nh·ªè nh·∫•t b·∫±ng Prim"""
    # X√¢y d·ª±ng danh s√°ch k·ªÅ
    adj = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        adj[u].append((v, w))
        adj[v].append((u, w))

    visited = [False] * (n + 1)
    mst = []
    total_weight = 0
    pq = []

    # B·∫Øt ƒë·∫ßu t·ª´ ƒë·ªânh 1
    heapq.heappush(pq, (0, 1, -1))  # (weight, current, parent)

    print("üîÑ Qu√° tr√¨nh Prim:")
    step = 1

    while pq and len(mst) < n - 1:
        weight, u, parent = heapq.heappop(pq)

        if visited[u]:
            continue

        visited[u] = True

        if parent != -1:
            mst.append((parent, u, weight))
            total_weight += weight
            print(f"B∆∞·ªõc {step}: Th√™m ({parent}, {u}, {weight}) - T·ªïng: {total_weight}")
            step += 1

        # Th√™m c√°c ƒë·ªânh k·ªÅ v√†o heap
        for v, w in adj[u]:
            if not visited[v]:
                heapq.heappush(pq, (w, v, u))

    return mst, total_weight

# Test B√†i 4
print("\nüìù TEST B√ÄI 4 - PRIM")
prim_mst, prim_weight = prim(n3, edges3)

print(f"\n‚úÖ K·∫æT QU·∫¢:")
print(f"S·ªë c·∫°nh: {len(prim_mst)}, T·ªïng tr·ªçng s·ªë: {prim_weight}")
print(f"C√°c c·∫°nh: {prim_mst}")

# V·∫Ω ƒë·ªì th·ªã B√†i 4
plt.figure(figsize=(10, 6))
nx.draw(G3, pos3, with_labels=True, node_color='lightblue',
        node_size=500, font_size=10, font_weight='bold')

# V·∫Ω t·∫•t c·∫£ c·∫°nh
nx.draw_networkx_edges(G3, pos3, alpha=0.3)

# V·∫Ω MST
prim_edges = [(u, v) for u, v, w in prim_mst]
nx.draw_networkx_edges(G3, pos3, edgelist=prim_edges,
                      edge_color='blue', width=2)

# V·∫Ω tr·ªçng s·ªë
nx.draw_networkx_edge_labels(G3, pos3, edge_labels=edge_labels)

plt.title("B√ÄI 4: PRIM - C√ÇY KHUNG NH·ªé NH·∫§T", fontsize=14, fontweight='bold')
plt.show()

# ============================================================================
# B√ÄI 5: M·∫†NG TRUY·ªÄN TH√îNG
# ============================================================================

print("\n" + "="*60)
print("B√ÄI 5: M·∫†NG TRUY·ªÄN TH√îNG")
print("="*60)

def analyze_network():
    """Ph√¢n t√≠ch m·∫°ng truy·ªÅn th√¥ng"""
    print("üéØ B√†i to√°n: X√°c ƒë·ªãnh ƒë∆∞·ªùng truy·ªÅn 'kh√¥ng ti·ªÅm nƒÉng'")

    # D·ªØ li·ªáu test case t·ª´ ƒë·ªÅ b√†i
    n5, m5, q5 = 3, 3, 2
    edges5 = [(1, 2, 1), (1, 3, 2), (2, 3, 3)]
    queries = [
        (3, [(2, 4), (3, 4)]),  # X√©t edge 3, thay ƒë·ªïi edge 2‚Üí4, edge 3‚Üí4
        (1, [(1, 4)])           # X√©t edge 1, thay ƒë·ªïi edge 1‚Üí4
    ]

    print(f"\nüìä ƒê·ªí TH·ªä BAN ƒê·∫¶U:")
    print(f"  {n5} ƒë·ªânh, {m5} c·∫°nh")
    for i, (u, v, w) in enumerate(edges5, 1):
        print(f"  Edge {i}: ({u}, {v}) - cost {w}")

    results = []

    for i, (k, changes) in enumerate(queries, 1):
        print(f"\nüîç TRUY V·∫§N {i}:")
        print(f"  X√©t edge {k}")
        print(f"  Thay ƒë·ªïi: {changes}")

        # T·∫°o b·∫£n sao v√† c·∫≠p nh·∫≠t
        temp_edges = edges5.copy()
        for t, c in changes:
            if 1 <= t <= len(temp_edges):
                u, v, old_w = temp_edges[t-1]
                temp_edges[t-1] = (u, v, c)
                print(f"    Edge {t}: cost {old_w} ‚Üí {c}")

        # T√¨m MST
        sorted_edges = sorted(temp_edges, key=lambda x: x[2])
        uf = UnionFind(n5)
        mst_edges = []

        for u, v, w in sorted_edges:
            if uf.union(u, v):
                mst_edges.append((u, v, w))

        # Ki·ªÉm tra edge k c√≥ trong MST kh√¥ng
        u_k, v_k, w_k = edges5[k-1]
        edge_in_mst = any(u == u_k and v == v_k for u, v, w in mst_edges)

        if edge_in_mst:
            print(f"  ‚ùå Edge {k} V·∫™N trong MST ‚Üí KH√îNG ti·ªÅm nƒÉng ‚Üí 'NO'")
            results.append("NO")
        else:
            print(f"  ‚úÖ Edge {k} KH√îNG trong MST ‚Üí TI·ªÄM NƒÇNG ‚Üí 'YES'")
            results.append("YES")

    return results

# Test B√†i 5
print("\nüìù TEST B√ÄI 5 - M·∫†NG TRUY·ªÄN TH√îNG")
network_results = analyze_network()

print(f"\n‚úÖ K·∫æT QU·∫¢:")
for i, result in enumerate(network_results, 1):
    print(f"  Truy v·∫•n {i}: {result}")

# ============================================================================
# T·ªîNG K·∫æT
# ============================================================================

print("\n" + "="*60)
print("üéä T·ªîNG K·∫æT 5 B√ÄI T·∫¨P")
print("="*60)

print(f"\nüìã B√ÄI 1 - C√¢y khung DFS:")
print(f"  S·ªë c·∫°nh: {len(dfs_tree)}")
print(f"  C√°c c·∫°nh: {dfs_tree}")

print(f"\nüìã B√ÄI 2 - C√¢y khung BFS:")
print(f"  S·ªë c·∫°nh: {len(bfs_tree)}")
print(f"  C√°c c·∫°nh: {bfs_tree}")

print(f"\nüìã B√ÄI 3 - Kruskal:")
print(f"  T·ªïng tr·ªçng s·ªë: {kruskal_weight}")
print(f"  C√°c c·∫°nh: {kruskal_mst}")

print(f"\nüìã B√ÄI 4 - Prim:")
print(f"  T·ªïng tr·ªçng s·ªë: {prim_weight}")
print(f"  C√°c c·∫°nh: {prim_mst}")

print(f"\nüìã B√ÄI 5 - M·∫°ng truy·ªÅn th√¥ng:")
for i, result in enumerate(network_results, 1):
    print(f"  Truy v·∫•n {i}: {result}")

print(f"\nüéØ SO S√ÅNH K·∫æT QU·∫¢:")
print(f"  DFS vs BFS: {'GI·ªêNG' if dfs_tree == bfs_tree else 'KH√ÅC'}")
print(f"  Kruskal vs Prim - Tr·ªçng s·ªë: {'B·∫∞NG' if kruskal_weight == prim_weight else 'KH√ÅC'}")

print(f"\nüìö T√ìM T·∫ÆT THU·∫¨T TO√ÅN:")
print(f"  ‚Ä¢ DFS/BFS: O(V+E) - ƒê·ªì th·ªã kh√¥ng tr·ªçng s·ªë")
print(f"  ‚Ä¢ Kruskal: O(E log E) - T·ªët cho ƒë·ªì th·ªã th∆∞a")
print(f"  ‚Ä¢ Prim: O(E log V) - T·ªët cho ƒë·ªì th·ªã d√†y")
print(f"  ‚Ä¢ Union-Find: C·∫•u tr√∫c hi·ªáu qu·∫£ cho Kruskal")

print("\n" + "="*60)
print("‚ú® K·∫æT TH√öC CH∆Ø∆†NG TR√åNH MINH H·ªåA ‚ú®")
print("="*60)
```
