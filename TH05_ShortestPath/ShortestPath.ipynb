```python
# -*- coding: utf-8 -*-
"""DUONG_DI_NGAN_NHAT.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1abc123
"""

# C√ÄI ƒê·∫∂T TH∆Ø VI·ªÜN
!pip install networkx matplotlib

import heapq
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from collections import deque

print("=== TH∆Ø VI·ªÜN ƒê√É S·∫¥N S√ÄNG ===")

# ƒê·ªí TH·ªä M·∫™U CHO T·∫§T C·∫¢ B√ÄI T·∫¨P
graph_data = {
    'nodes': [1, 2, 3, 4, 5, 6],
    'edges': [
        (1, 2, 1), (1, 6, 10),
        (2, 3, 2),
        (3, 4, 20), (3, 6, 3),
        (6, 5, 4),
        (5, 4, 5)
    ]
}

def visualize_graph(graph_data, path=None, title="ƒê·ªì th·ªã"):
    """V·∫Ω ƒë·ªì th·ªã v·ªõi ƒë∆∞·ªùng ƒëi ƒë∆∞·ª£c t√¥ s√°ng"""
    G = nx.DiGraph()

    # Th√™m c√°c ƒë·ªânh v√† c·∫°nh
    for node in graph_data['nodes']:
        G.add_node(node)

    for u, v, w in graph_data['edges']:
        G.add_edge(u, v, weight=w)

    pos = nx.spring_layout(G, seed=42)

    plt.figure(figsize=(12, 8))

    # V·∫Ω t·∫•t c·∫£ c√°c c·∫°nh
    nx.draw_networkx_edges(G, pos, alpha=0.3, width=2, arrows=True)

    # V·∫Ω ƒë∆∞·ªùng ƒëi ƒë∆∞·ª£c t√¥ s√°ng (n·∫øu c√≥)
    if path:
        path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
        nx.draw_networkx_edges(G, pos, edgelist=path_edges,
                              edge_color='red', width=3, arrows=True)

    # V·∫Ω c√°c ƒë·ªânh
    nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                          node_size=800, alpha=0.7)

    # V·∫Ω nh√£n ƒë·ªânh v√† tr·ªçng s·ªë
    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')
    edge_labels = {(u, v): f'{w}' for u, v, w in graph_data['edges']}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

print("=== B√ÄI 1: DIJKSTRA - ƒê∆Ø·ªúNG ƒêI NG·∫ÆN NH·∫§T ===")

def dijkstra_demo(graph_data, start, end):
    """Minh h·ªça thu·∫≠t to√°n Dijkstra"""
    # X√¢y d·ª±ng ƒë·ªì th·ªã
    graph = {}
    for u, v, w in graph_data['edges']:
        if u not in graph:
            graph[u] = []
        graph[u].append((v, w))

    # Kh·ªüi t·∫°o
    dist = {node: float('inf') for node in graph_data['nodes']}
    dist[start] = 0
    parent = {node: None for node in graph_data['nodes']}
    visited = set()

    # Priority queue
    pq = [(0, start)]

    print(f"üéØ DIJKSTRA T·ª™ {start} ƒê·∫æN {end}")
    print(f"Kh·ªüi t·∫°o: dist[{start}] = 0")

    step = 1
    while pq:
        current_dist, u = heapq.heappop(pq)

        if u in visited:
            continue

        visited.add(u)
        print(f"\nüîç B∆∞·ªõc {step}: X√©t ƒë·ªânh {u} (kho·∫£ng c√°ch = {current_dist})")
        step += 1

        if u == end:
            print(f"üéâ ƒê√É T√åM TH·∫§Y ƒê√çCH {end}!")
            break

        if u in graph:
            for v, weight in graph[u]:
                new_dist = current_dist + weight
                print(f"   ‚Üí X√©t ({u}‚Üí{v}) v·ªõi tr·ªçng s·ªë {weight}")

                if new_dist < dist[v]:
                    dist[v] = new_dist
                    parent[v] = u
                    heapq.heappush(pq, (new_dist, v))
                    print(f"     ‚úÖ C·∫≠p nh·∫≠t dist[{v}] = {new_dist}, parent[{v}] = {u}")
                else:
                    print(f"     ‚ùå Gi·ªØ nguy√™n dist[{v}] = {dist[v]}")

        print(f"   Priority queue: {pq}")

    # Truy v·∫øt ƒë∆∞·ªùng ƒëi
    if dist[end] != float('inf'):
        path = []
        node = end
        while node is not None:
            path.append(node)
            node = parent[node]
        path.reverse()

        print(f"\nüìä K·∫æT QU·∫¢:")
        print(f"   ƒê·ªô d√†i ƒë∆∞·ªùng ƒëi: {dist[end]}")
        print(f"   ƒê∆∞·ªùng ƒëi: {' ‚Üí '.join(map(str, path))}")

        # Visualize
        visualize_graph(graph_data, path,
                       f"B√†i 1: Dijkstra - ƒê∆∞·ªùng ƒëi {start}‚Üí{end}\nƒê·ªô d√†i: {dist[end]}")

        return path, dist[end]
    else:
        print(f"üí• KH√îNG C√ì ƒê∆Ø·ªúNG ƒêI T·ª™ {start} ƒê·∫æN {end}")
        return None, float('inf')

# CH·∫†Y B√ÄI 1
path1, dist1 = dijkstra_demo(graph_data, 1, 4)

print("\n" + "="*60)
print("=== B√ÄI 2: ƒê∆Ø·ªúNG ƒêI QUA ƒê·ªàNH TRUNG GIAN ===")

def dijkstra_via_intermediate(graph_data, start, end, via):
    """T√¨m ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t qua ƒë·ªânh trung gian"""
    print(f"üéØ T√åM ƒê∆Ø·ªúNG ƒêI {start} ‚Üí {via} ‚Üí {end}")

    # T√¨m ƒë∆∞·ªùng t·ª´ start ƒë·∫øn via
    print(f"\n1. T√åM ƒê∆Ø·ªúNG {start} ‚Üí {via}")
    path1, dist1 = dijkstra_demo(graph_data, start, via)

    if path1 is None:
        print(f"üí• KH√îNG C√ì ƒê∆Ø·ªúNG ƒêI T·ª™ {start} ƒê·∫æN {via}")
        return None, float('inf')

    # T√¨m ƒë∆∞·ªùng t·ª´ via ƒë·∫øn end
    print(f"\n2. T√åM ƒê∆Ø·ªúNG {via} ‚Üí {end}")
    path2, dist2 = dijkstra_demo(graph_data, via, end)

    if path2 is None:
        print(f"üí• KH√îNG C√ì ƒê∆Ø·ªúNG ƒêI T·ª™ {via} ƒê·∫æN {end}")
        return None, float('inf')

    # K·∫øt h·ª£p ƒë∆∞·ªùng ƒëi
    total_dist = dist1 + dist2
    combined_path = path1[:-1] + path2  # Tr√°nh l·∫∑p ƒë·ªânh via

    print(f"\nüìä K·∫æT QU·∫¢ CU·ªêI C√ôNG:")
    print(f"   ƒê·ªô d√†i {start}‚Üí{via}: {dist1}")
    print(f"   ƒê·ªô d√†i {via}‚Üí{end}: {dist2}")
    print(f"   T·ªïng ƒë·ªô d√†i: {total_dist}")
    print(f"   ƒê∆∞·ªùng ƒëi: {' ‚Üí '.join(map(str, combined_path))}")

    # Visualize
    visualize_graph(graph_data, combined_path,
                   f"B√†i 2: ƒê∆∞·ªùng ƒëi {start}‚Üí{via}‚Üí{end}\nT·ªïng ƒë·ªô d√†i: {total_dist}")

    return combined_path, total_dist

# CH·∫†Y B√ÄI 2
path2, dist2 = dijkstra_via_intermediate(graph_data, 1, 4, 3)

print("\n" + "="*60)
print("=== B√ÄI 3: FLOYD-WARSHALL - M·ªåI C·∫∂P ƒê·ªàNH ===")

def floyd_warshall_demo(graph_data):
    """Minh h·ªça thu·∫≠t to√°n Floyd-Warshall"""
    nodes = sorted(graph_data['nodes'])
    n = len(nodes)
    node_to_idx = {node: i for i, node in enumerate(nodes)}
    idx_to_node = {i: node for i, node in enumerate(nodes)}

    # Kh·ªüi t·∫°o ma tr·∫≠n kho·∫£ng c√°ch
    dist = np.full((n, n), float('inf'))
    for i in range(n):
        dist[i][i] = 0

    # ƒêi·ªÅn c√°c c·∫°nh
    for u, v, w in graph_data['edges']:
        i, j = node_to_idx[u], node_to_idx[v]
        dist[i][j] = w
        # ƒê·ªì th·ªã v√¥ h∆∞·ªõng
        dist[j][i] = w

    print("üéØ FLOYD-WARSHALL - MA TR·∫¨N KHO·∫¢NG C√ÅCH")
    print("\nüìä Ma tr·∫≠n ban ƒë·∫ßu:")
    print_matrix(dist, idx_to_node)

    # Thu·∫≠t to√°n Floyd-Warshall
    for k in range(n):
        print(f"\nüåÄ B∆∞·ªõc {k+1}: X√©t ƒë·ªânh trung gian {idx_to_node[k]}")

        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    old_val = dist[i][j]
                    dist[i][j] = dist[i][k] + dist[k][j]
                    print(f"   dist[{idx_to_node[i]}][{idx_to_node[j]}] = " +
                          f"min({old_val}, {dist[i][k]} + {dist[k][j]}) = {dist[i][j]}")

    print(f"\nüìä MA TR·∫¨N K·∫æT QU·∫¢:")
    print_matrix(dist, idx_to_node)

    return dist, idx_to_node

def print_matrix(matrix, idx_to_node):
    """In ma tr·∫≠n kho·∫£ng c√°ch"""
    n = len(matrix)

    # In header
    print("    " + " ".join(f"{idx_to_node[i]:>4}" for i in range(n)))

    for i in range(n):
        row = [f"{idx_to_node[i]:>2}"]
        for j in range(n):
            if matrix[i][j] == float('inf'):
                row.append("  ‚àû")
            else:
                row.append(f"{matrix[i][j]:>3}")
        print(" ".join(row))

# CH·∫†Y B√ÄI 3
dist_matrix, idx_map = floyd_warshall_demo(graph_data)

print("\n" + "="*60)
print("=== B√ÄI 4: DIJKSTRA V·ªöI PRIORITY QUEUE ===")

def dijkstra_priority_queue_demo(graph_data, start, end):
    """Minh h·ªça Dijkstra v·ªõi priority queue chi ti·∫øt"""
    graph = {}
    for u, v, w in graph_data['edges']:
        if u not in graph:
            graph[u] = []
        graph[u].append((v, w))

    dist = {node: float('inf') for node in graph_data['nodes']}
    dist[start] = 0
    parent = {node: None for node in graph_data['nodes']}
    visited = set()

    pq = [(0, start)]
    operations = []

    print(f"üéØ DIJKSTRA V·ªöI PRIORITY QUEUE - {start}‚Üí{end}")
    print(f"Kh·ªüi t·∫°o: pq = {pq}")

    step = 1
    while pq:
        current_dist, u = heapq.heappop(pq)
        operations.append(f"B∆∞·ªõc {step}: L·∫•y ({current_dist}, {u}) t·ª´ pq")

        if u in visited:
            operations.append(f"  ‚Üí B·ªè qua {u} (ƒë√£ visited)")
            continue

        visited.add(u)
        operations.append(f"  ‚Üí X·ª≠ l√Ω ƒë·ªânh {u}")

        if u == end:
            operations.append(f"  üéâ T√åM TH·∫§Y ƒê√çCH {end}!")
            break

        if u in graph:
            for v, weight in graph[u]:
                new_dist = current_dist + weight
                operations.append(f"    X√©t {u}‚Üí{v} (w={weight})")

                if new_dist < dist[v]:
                    old_dist = dist[v]
                    dist[v] = new_dist
                    parent[v] = u
                    heapq.heappush(pq, (new_dist, v))
                    operations.append(f"      ‚úÖ C·∫≠p nh·∫≠t: dist[{v}] {old_dist}‚Üí{new_dist}")
                    operations.append(f"      Th√™m ({new_dist}, {v}) v√†o pq")
                else:
                    operations.append(f"      ‚ùå Gi·ªØ nguy√™n: dist[{v}] = {dist[v]}")

        operations.append(f"    pq hi·ªán t·∫°i: {pq}")
        step += 1

    # In chi ti·∫øt c√°c b∆∞·ªõc
    print("\nüìù CHI TI·∫æT C√ÅC B∆Ø·ªöC:")
    for op in operations:
        print(f"   {op}")

    # K·∫øt qu·∫£
    if dist[end] != float('inf'):
        path = []
        node = end
        while node is not None:
            path.append(node)
            node = parent[node]
        path.reverse()

        print(f"\nüìä K·∫æT QU·∫¢:")
        print(f"   ƒê·ªô d√†i: {dist[end]}")
        print(f"   ƒê∆∞·ªùng ƒëi: {' ‚Üí '.join(map(str, path))}")
        print(f"   S·ªë ƒë·ªânh: {len(path)}")

        visualize_graph(graph_data, path,
                       f"B√†i 4: Dijkstra PQ - {start}‚Üí{end}\nƒê·ªô d√†i: {dist[end]}, S·ªë ƒë·ªânh: {len(path)}")

        return path, dist[end]
    else:
        print(f"üí• KH√îNG C√ì ƒê∆Ø·ªúNG ƒêI")
        return None, float('inf')

# CH·∫†Y B√ÄI 4
path4, dist4 = dijkstra_priority_queue_demo(graph_data, 1, 4)

print("\n" + "="*60)
print("üéä T·ªîNG K·∫æT 4 B√ÄI T·∫¨P ƒê∆Ø·ªúNG ƒêI NG·∫ÆN NH·∫§T")
print("="*60)

print(f"\nüìã B√ÄI 1 - Dijkstra c∆° b·∫£n:")
print(f"   {1}‚Üí{4}: ƒê·ªô d√†i = {dist1}, ƒê∆∞·ªùng ƒëi = {' ‚Üí '.join(map(str, path1))}")

print(f"\nüìã B√ÄI 2 - Qua ƒë·ªânh trung gian:")
print(f"   {1}‚Üí{3}‚Üí{4}: ƒê·ªô d√†i = {dist2}, ƒê∆∞·ªùng ƒëi = {' ‚Üí '.join(map(str, path2))}")

print(f"\nüìã B√ÄI 3 - Floyd-Warshall:")
print(f"   ƒê√£ t√≠nh kho·∫£ng c√°ch gi·ªØa m·ªçi c·∫∑p ƒë·ªânh")

print(f"\nüìã B√ÄI 4 - Dijkstra Priority Queue:")
print(f"   {1}‚Üí{4}: ƒê·ªô d√†i = {dist4}, S·ªë ƒë·ªânh = {len(path4)}")

print(f"\nüéØ SO S√ÅNH K·∫æT QU·∫¢:")
print(f"   Dijkstra c∆° b·∫£n vs Priority Queue: {'GI·ªêNG' if dist1 == dist4 else 'KH√ÅC'}")

print(f"\nüìö T√ìM T·∫ÆT THU·∫¨T TO√ÅN:")
print(f"   ‚Ä¢ Dijkstra: ƒê∆∞·ªùng ƒëi ng·∫Øn nh·∫•t t·ª´ 1 ƒë·ªânh, O((V+E)logV)")
print(f"   ‚Ä¢ Floyd-Warshall: M·ªçi c·∫∑p ƒë·ªânh, O(V¬≥)")
print(f"   ‚Ä¢ Priority Queue: T·ªëi ∆∞u cho ƒë·ªì th·ªã th∆∞a")
print(f"   ‚Ä¢ ƒê·ªânh trung gian: Chia b√†i to√°n th√†nh 2 ph·∫ßn")

print("\n" + "="*60)
print("‚ú® K·∫æT TH√öC CH∆Ø∆†NG TR√åNH MINH H·ªåA ‚ú®")
print("="*60)
```
