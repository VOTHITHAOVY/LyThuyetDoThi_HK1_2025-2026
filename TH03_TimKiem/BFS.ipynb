```python
# -*- coding: utf-8 -*-
"""BFS_DEMO_ALL_IN_ONE.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1abc123
"""

# C√ÄI ƒê·∫∂T TH∆Ø VI·ªÜN (ch·∫°y cell n√†y ƒë·∫ßu ti√™n)
!pip install networkx matplotlib

import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
import numpy as np

print("=== TH∆Ø VI·ªÜN ƒê√É S·∫¥N S√ÄNG ===")

# ƒê·ªí TH·ªä M·∫™U CHO T·∫§T C·∫¢ B√ÄI T·∫¨P
graph = {
    1: [2],
    2: [1, 5, 4],
    3: [7],
    4: [2, 8],
    5: [2],
    6: [],
    7: [3, 2],
    8: [4]
}

def visualize_graph(highlight_nodes=None, highlight_edges=None, title="ƒê·ªì th·ªã"):
    """H√†m v·∫Ω ƒë·ªì th·ªã v·ªõi c√°c ƒë·ªânh v√† c·∫°nh ƒë∆∞·ª£c t√¥ s√°ng"""
    G = nx.Graph()

    # Th√™m c√°c ƒë·ªânh v√† c·∫°nh
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            if neighbor > node:  # Tr√°nh th√™m c·∫°nh tr√πng
                G.add_edge(node, neighbor)

    # V·ªã tr√≠ c√°c ƒë·ªânh
    pos = {
        1: (0, 1), 2: (1, 1), 3: (2, 2), 4: (1, 0),
        5: (2, 1), 6: (3, 0), 7: (1.5, 1.5), 8: (1, -1)
    }

    plt.figure(figsize=(10, 8))

    # V·∫Ω t·∫•t c·∫£ c√°c c·∫°nh (m√†u x√°m)
    nx.draw_networkx_edges(G, pos, alpha=0.3, width=2)

    # V·∫Ω c√°c c·∫°nh ƒë∆∞·ª£c t√¥ s√°ng (n·∫øu c√≥)
    if highlight_edges:
        nx.draw_networkx_edges(G, pos, edgelist=highlight_edges,
                              edge_color='red', width=3)

    # V·∫Ω t·∫•t c·∫£ c√°c ƒë·ªânh
    nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                          node_size=800, alpha=0.7)

    # V·∫Ω c√°c ƒë·ªânh ƒë∆∞·ª£c t√¥ s√°ng (n·∫øu c√≥)
    if highlight_nodes:
        nx.draw_networkx_nodes(G, pos, nodelist=highlight_nodes,
                              node_color='red', node_size=800)

    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

print("=== B√ÄI 1: C√ÅC ƒê·ªàNH LI√äN TH√îNG V·ªöI x ===")

def bfs_connected_component_demo(graph, start):
    """Minh h·ªça BFS t√¨m c√°c ƒë·ªânh li√™n th√¥ng"""
    visited = set()
    queue = deque([start])
    visited.add(start)
    steps = []
    visited_order = [start]

    print(f"üéØ BFS T·ª™ ƒê·ªàNH {start}")
    print(f"B·∫Øt ƒë·∫ßu t·ª´ ƒë·ªânh: {start}")
    print(f"Queue kh·ªüi t·∫°o: {[start]}")

    step_count = 1
    while queue:
        current = queue.popleft()
        steps.append(f"B∆∞·ªõc {step_count}: L·∫•y ƒë·ªânh {current} t·ª´ queue")
        step_count += 1

        print(f"\nüîç B∆∞·ªõc {step_count-1}: X√©t ƒë·ªânh {current}")
        print(f"   C√°c ƒë·ªânh k·ªÅ: {graph[current]}")

        neighbors_added = []
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                visited_order.append(neighbor)
                neighbors_added.append(neighbor)
                print(f"   ‚úÖ Th√™m {neighbor} v√†o queue")
            else:
                print(f"   ‚ùå B·ªè qua {neighbor} (ƒë√£ visited)")

        print(f"   Queue hi·ªán t·∫°i: {list(queue)}")
        print(f"   ƒê√£ visited: {sorted(visited)}")

    print(f"\nüéâ K·∫æT QU·∫¢ B√ÄI 1:")
    print(f"   T·ª´ ƒë·ªânh {start} c√≥ th·ªÉ ƒë·∫øn {len(visited)} ƒë·ªânh")
    print(f"   Danh s√°ch: {sorted(visited)}")

    # Visualize k·∫øt qu·∫£
    visualize_graph(highlight_nodes=visited,
                   title=f"B√†i 1: C√°c ƒë·ªânh li√™n th√¥ng v·ªõi {start}")

    return visited

# CH·∫†Y B√ÄI 1
visited_nodes = bfs_connected_component_demo(graph, 7)

print("\n" + "="*50)
print("=== B√ÄI 2: T√åM ƒê∆Ø·ªúNG ƒêI B·∫∞NG BFS ===")

def bfs_path_finding_demo(graph, start, end):
    """Minh h·ªça BFS t√¨m ƒë∆∞·ªùng ƒëi ng·∫Øn nh·∫•t"""
    visited = set()
    queue = deque([start])
    visited.add(start)
    parent = {start: None}

    print(f"üéØ T√åM ƒê∆Ø·ªúNG ƒêI T·ª™ {start} ƒê·∫æN {end}")
    print(f"Kh·ªüi t·∫°o t·ª´ ƒë·ªânh: {start}")

    found = False
    step_count = 1

    while queue and not found:
        current = queue.popleft()

        print(f"\nüîç B∆∞·ªõc {step_count}: X√©t ƒë·ªânh {current}")
        step_count += 1

        if current == end:
            print(f"üéâ T√åM TH·∫§Y ƒê√çCH {end}!")
            found = True
            break

        neighbors_added = []
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                queue.append(neighbor)
                neighbors_added.append(neighbor)
                print(f"   ‚úÖ Th√™m {neighbor} (t·ª´ {current})")

        print(f"   Queue: {list(queue)}")

    # Truy v·∫øt ƒë∆∞·ªùng ƒëi
    if found:
        path = []
        node = end
        while node is not None:
            path.append(node)
            node = parent[node]
        path.reverse()

        print(f"\nüó∫Ô∏è ƒê∆Ø·ªúNG ƒêI T√åM ƒê∆Ø·ª¢C:")
        print(f"   {' ‚Üí '.join(map(str, path))}")
        print(f"   ƒê·ªô d√†i: {len(path)} ƒë·ªânh")

        # Visualize ƒë∆∞·ªùng ƒëi
        path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
        visualize_graph(highlight_nodes=path, highlight_edges=path_edges,
                      title=f"B√†i 2: ƒê∆∞·ªùng ƒëi {start} ‚Üí {end}")

        return path
    else:
        print(f"üí• KH√îNG T√åM TH·∫§Y ƒê∆Ø·ªúNG ƒêI")
        return None

# CH·∫†Y B√ÄI 2
path = bfs_path_finding_demo(graph, 3, 8)

print("\n" + "="*50)
print("=== B√ÄI 3: TH√ÄNH PH·∫¶N LI√äN TH√îNG ===")

def find_connected_components_demo(graph):
    """Minh h·ªça t√¨m t·∫•t c·∫£ th√†nh ph·∫ßn li√™n th√¥ng"""
    visited = set()
    components = []

    print("üîç T√åM C√ÅC TH√ÄNH PH·∫¶N LI√äN TH√îNG")

    component_count = 0
    for node in sorted(graph.keys()):
        if node not in visited:
            component_count += 1
            print(f"\n--- Th√†nh ph·∫ßn li√™n th√¥ng #{component_count} ---")

            # BFS cho th√†nh ph·∫ßn m·ªõi
            component = []
            queue = deque([node])
            visited.add(node)
            component.append(node)

            print(f"B·∫Øt ƒë·∫ßu BFS t·ª´ ƒë·ªânh {node}")

            while queue:
                current = queue.popleft()

                for neighbor in graph[current]:
                    if neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
                        component.append(neighbor)
                        print(f"   + Th√™m {neighbor}")

            components.append(sorted(component))
            print(f"Th√†nh ph·∫ßn {component_count}: {sorted(component)}")

    print(f"\nüìä T·ªîNG K·∫æT B√ÄI 3:")
    print(f"   S·ªë th√†nh ph·∫ßn li√™n th√¥ng: {len(components)}")
    for i, comp in enumerate(components, 1):
        print(f"   Th√†nh ph·∫ßn {i}: {comp}")

    # Visualize c√°c th√†nh ph·∫ßn li√™n th√¥ng
    plt.figure(figsize=(12, 8))
    G = nx.Graph()

    for node, neighbors in graph.items():
        for neighbor in neighbors:
            if neighbor > node:
                G.add_edge(node, neighbor)

    pos = {
        1: (0, 1), 2: (1, 1), 3: (2, 2), 4: (1, 0),
        5: (2, 1), 6: (3, 0), 7: (1.5, 1.5), 8: (1, -1)
    }

    # M√†u cho c√°c th√†nh ph·∫ßn
    colors = ['red', 'blue', 'green', 'orange', 'purple']

    for i, component in enumerate(components):
        nx.draw_networkx_nodes(G, pos, nodelist=component,
                              node_color=colors[i % len(colors)],
                              node_size=800, alpha=0.7)

    nx.draw_networkx_edges(G, pos, alpha=0.6, width=2)
    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

    plt.title("B√†i 3: C√°c th√†nh ph·∫ßn li√™n th√¥ng", fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

    return components

# CH·∫†Y B√ÄI 3
components = find_connected_components_demo(graph)

print("\n" + "="*50)
print("=== B√ÄI 4: C·∫†NH C·∫¶U V√Ä ƒê·ªàNH KH·ªöP ===")

def count_components(graph, remove_node=None, remove_edge=None):
    """ƒê·∫øm s·ªë th√†nh ph·∫ßn li√™n th√¥ng"""
    temp_graph = {node: neighbors[:] for node, neighbors in graph.items()}

    # X√≥a ƒë·ªânh n·∫øu c√≥
    if remove_node:
        temp_graph[remove_node] = []
        for node in temp_graph:
            if remove_node in temp_graph[node]:
                temp_graph[node].remove(remove_node)

    # X√≥a c·∫°nh n·∫øu c√≥
    if remove_edge:
        u, v = remove_edge
        if v in temp_graph[u]:
            temp_graph[u].remove(v)
        if u in temp_graph[v]:
            temp_graph[v].remove(u)

    # ƒê·∫øm th√†nh ph·∫ßn li√™n th√¥ng
    visited = set()
    count = 0

    for node in temp_graph:
        if node not in visited and temp_graph[node]:
            queue = deque([node])
            visited.add(node)

            while queue:
                current = queue.popleft()
                for neighbor in temp_graph[current]:
                    if neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
            count += 1

    return count

def check_bridge_and_articulation_demo(graph, test_edge=(2, 4), test_node=7):
    """Minh h·ªça ki·ªÉm tra c·∫°nh c·∫ßu v√† ƒë·ªânh kh·ªõp"""
    print("üîç KI·ªÇM TRA C·∫†NH C·∫¶U V√Ä ƒê·ªàNH KH·ªöP")

    # Ki·ªÉm tra c·∫°nh c·∫ßu
    print(f"\n1. KI·ªÇM TRA C·∫†NH {test_edge}:")
    original_components = count_components(graph)
    components_after_edge = count_components(graph, remove_edge=test_edge)

    print(f"   - TPLT ban ƒë·∫ßu: {original_components}")
    print(f"   - TPLT sau x√≥a c·∫°nh: {components_after_edge}")

    is_bridge = components_after_edge > original_components
    result_bridge = "‚úÖ L√Ä C·∫†NH C·∫¶U" if is_bridge else "‚ùå KH√îNG PH·∫¢I C·∫†NH C·∫¶U"
    print(f"   - K·∫øt lu·∫≠n: {result_bridge}")

    # Ki·ªÉm tra ƒë·ªânh kh·ªõp
    print(f"\n2. KI·ªÇM TRA ƒê·ªàNH {test_node}:")
    components_after_node = count_components(graph, remove_node=test_node)

    print(f"   - TPLT ban ƒë·∫ßu: {original_components}")
    print(f"   - TPLT sau x√≥a ƒë·ªânh: {components_after_node}")

    is_articulation = components_after_node > original_components
    result_articulation = "‚úÖ L√Ä ƒê·ªàNH KH·ªöP" if is_articulation else "‚ùå KH√îNG PH·∫¢I ƒê·ªàNH KH·ªöP"
    print(f"   - K·∫øt lu·∫≠n: {result_articulation}")

    # Visualize
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            if neighbor > node:
                G.add_edge(node, neighbor)

    pos = {
        1: (0, 1), 2: (1, 1), 3: (2, 2), 4: (1, 0),
        5: (2, 1), 6: (3, 0), 7: (1.5, 1.5), 8: (1, -1)
    }

    # Subplot 1: C·∫°nh c·∫ßu
    nx.draw(G, pos, ax=ax1, with_labels=True, node_color='lightblue',
            node_size=800, font_weight='bold')

    # T√¥ ƒë·ªè c·∫°nh ƒëang ki·ªÉm tra
    if test_edge in G.edges() or (test_edge[1], test_edge[0]) in G.edges():
        edge_to_highlight = test_edge if test_edge in G.edges() else (test_edge[1], test_edge[0])
        nx.draw_networkx_edges(G, pos, edgelist=[edge_to_highlight],
                              edge_color='red', width=3, ax=ax1)

    ax1.set_title(f"C·∫°nh {test_edge}\n{result_bridge}",
                 fontweight='bold', color='red' if is_bridge else 'green')

    # Subplot 2: ƒê·ªânh kh·ªõp
    nx.draw(G, pos, ax=ax2, with_labels=True, node_color='lightblue',
            node_size=800, font_weight='bold')

    # T√¥ ƒë·ªè ƒë·ªânh ƒëang ki·ªÉm tra
    nx.draw_networkx_nodes(G, pos, nodelist=[test_node],
                          node_color='red', node_size=800, ax=ax2)

    ax2.set_title(f"ƒê·ªânh {test_node}\n{result_articulation}",
                 fontweight='bold', color='red' if is_articulation else 'green')

    plt.tight_layout()
    plt.show()

    return is_bridge, is_articulation

# CH·∫†Y B√ÄI 4
is_bridge, is_articulation = check_bridge_and_articulation_demo(graph)

print("\n" + "="*60)
print("üéä T·ªîNG K·∫æT 4 B√ÄI T·∫¨P BFS")
print("="*60)

print(f"\nüìã B√ÄI 1 - Li√™n th√¥ng v·ªõi ƒë·ªânh 7:")
print(f"   K·∫øt qu·∫£: {sorted(visited_nodes)}")

print(f"\nüìã B√ÄI 2 - ƒê∆∞·ªùng ƒëi t·ª´ 3 ƒë·∫øn 8:")
print(f"   K·∫øt qu·∫£: {' ‚Üí '.join(map(str, path))}")

print(f"\nüìã B√ÄI 3 - Th√†nh ph·∫ßn li√™n th√¥ng:")
for i, comp in enumerate(components, 1):
    print(f"   Th√†nh ph·∫ßn {i}: {comp}")

print(f"\nüìã B√ÄI 4 - C·∫°nh c·∫ßu v√† ƒë·ªânh kh·ªõp:")
print(f"   C·∫°nh (2,4): {'L√Ä C·∫¶U' if is_bridge else 'KH√îNG PH·∫¢I C·∫¶U'}")
print(f"   ƒê·ªânh 7: {'L√Ä KH·ªöP' if is_articulation else 'KH√îNG PH·∫¢I KH·ªöP'}")

print("\n" + "="*60)
print("‚ú® K·∫æT TH√öC CH∆Ø∆†NG TR√åNH MINH H·ªåA ‚ú®")
print("="*60)
```
