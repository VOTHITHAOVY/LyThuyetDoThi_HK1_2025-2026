```python
# -*- coding: utf-8 -*-
"""DFS_DEMO_ALL_IN_ONE.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1abc123
"""

# C√ÄI ƒê·∫∂T TH∆Ø VI·ªÜN (ch·∫°y cell n√†y ƒë·∫ßu ti√™n)
!pip install networkx matplotlib

import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
import numpy as np

print("=== TH∆Ø VI·ªÜN ƒê√É S·∫¥N S√ÄNG ===")

# ƒê·ªí TH·ªä M·∫™U CHO T·∫§T C·∫¢ B√ÄI T·∫¨P
graph = {
    1: [2],
    2: [1, 5, 4],
    3: [7],
    4: [2, 8],
    5: [2],
    6: [],
    7: [3, 2],
    8: [4]
}

def visualize_graph(highlight_nodes=None, highlight_edges=None, title="ƒê·ªì th·ªã", bipartite_colors=None):
    """H√†m v·∫Ω ƒë·ªì th·ªã v·ªõi c√°c ƒë·ªânh v√† c·∫°nh ƒë∆∞·ª£c t√¥ s√°ng"""
    G = nx.Graph()

    # Th√™m c√°c ƒë·ªânh v√† c·∫°nh
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            if neighbor > node:  # Tr√°nh th√™m c·∫°nh tr√πng
                G.add_edge(node, neighbor)

    # V·ªã tr√≠ c√°c ƒë·ªânh
    pos = {
        1: (0, 1), 2: (1, 1), 3: (2, 2), 4: (1, 0),
        5: (2, 1), 6: (3, 0), 7: (1.5, 1.5), 8: (1, -1)
    }

    plt.figure(figsize=(10, 8))

    # V·∫Ω t·∫•t c·∫£ c√°c c·∫°nh (m√†u x√°m)
    nx.draw_networkx_edges(G, pos, alpha=0.3, width=2)

    # V·∫Ω c√°c c·∫°nh ƒë∆∞·ª£c t√¥ s√°ng (n·∫øu c√≥)
    if highlight_edges:
        nx.draw_networkx_edges(G, pos, edgelist=highlight_edges,
                              edge_color='red', width=3)

    # V·∫Ω c√°c ƒë·ªânh v·ªõi m√†u ph√¢n ƒë√¥i (n·∫øu c√≥)
    if bipartite_colors:
        color_map = []
        for node in G.nodes():
            color_map.append(bipartite_colors.get(node, 'lightblue'))
        nx.draw_networkx_nodes(G, pos, node_color=color_map,
                              node_size=800, alpha=0.7)
    else:
        # V·∫Ω t·∫•t c·∫£ c√°c ƒë·ªânh
        nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                              node_size=800, alpha=0.7)

    # V·∫Ω c√°c ƒë·ªânh ƒë∆∞·ª£c t√¥ s√°ng (n·∫øu c√≥)
    if highlight_nodes:
        nx.draw_networkx_nodes(G, pos, nodelist=highlight_nodes,
                              node_color='red', node_size=800)

    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

print("=== B√ÄI 1: C√ÅC ƒê·ªàNH LI√äN TH√îNG V·ªöI x (DFS) ===")

def dfs_connected_component_demo(graph, start):
    """Minh h·ªça DFS t√¨m c√°c ƒë·ªânh li√™n th√¥ng"""
    visited = set()
    stack = [start]
    visited.add(start)
    steps = []
    visited_order = [start]

    print(f"üéØ DFS T·ª™ ƒê·ªàNH {start}")
    print(f"B·∫Øt ƒë·∫ßu t·ª´ ƒë·ªânh: {start}")
    print(f"Stack kh·ªüi t·∫°o: {[start]}")

    step_count = 1
    while stack:
        current = stack.pop()
        steps.append(f"B∆∞·ªõc {step_count}: L·∫•y ƒë·ªânh {current} t·ª´ stack")
        step_count += 1

        print(f"\nüîç B∆∞·ªõc {step_count-1}: X√©t ƒë·ªânh {current}")
        print(f"   C√°c ƒë·ªânh k·ªÅ: {graph[current]}")

        neighbors_added = []
        # Duy·ªát theo th·ª© t·ª± ng∆∞·ª£c ƒë·ªÉ gi·ªëng DFS ƒë·ªá quy
        for neighbor in reversed(graph[current]):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
                visited_order.append(neighbor)
                neighbors_added.append(neighbor)
                print(f"   ‚úÖ Th√™m {neighbor} v√†o stack")
            else:
                print(f"   ‚ùå B·ªè qua {neighbor} (ƒë√£ visited)")

        print(f"   Stack hi·ªán t·∫°i: {stack}")
        print(f"   ƒê√£ visited: {sorted(visited)}")

    print(f"\nüéâ K·∫æT QU·∫¢ B√ÄI 1:")
    print(f"   T·ª´ ƒë·ªânh {start} c√≥ th·ªÉ ƒë·∫øn {len(visited)} ƒë·ªânh")
    print(f"   Danh s√°ch: {sorted(visited)}")

    # Visualize k·∫øt qu·∫£
    visualize_graph(highlight_nodes=visited,
                   title=f"B√†i 1: C√°c ƒë·ªânh li√™n th√¥ng v·ªõi {start} (DFS)")

    return visited

# CH·∫†Y B√ÄI 1
visited_nodes = dfs_connected_component_demo(graph, 7)

print("\n" + "="*50)
print("=== B√ÄI 2: T√åM ƒê∆Ø·ªúNG ƒêI B·∫∞NG DFS ===")

def dfs_path_finding_demo(graph, start, end):
    """Minh h·ªça DFS t√¨m ƒë∆∞·ªùng ƒëi"""
    visited = set()
    stack = [start]
    visited.add(start)
    parent = {start: None}

    print(f"üéØ T√åM ƒê∆Ø·ªúNG ƒêI T·ª™ {start} ƒê·∫æN {end} (DFS)")
    print(f"Kh·ªüi t·∫°o t·ª´ ƒë·ªânh: {start}")

    found = False
    step_count = 1

    while stack and not found:
        current = stack.pop()

        print(f"\nüîç B∆∞·ªõc {step_count}: X√©t ƒë·ªânh {current}")
        step_count += 1

        if current == end:
            print(f"üéâ T√åM TH·∫§Y ƒê√çCH {end}!")
            found = True
            break

        neighbors_added = []
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                stack.append(neighbor)
                neighbors_added.append(neighbor)
                print(f"   ‚úÖ Th√™m {neighbor} (t·ª´ {current})")

        print(f"   Stack: {stack}")

    # Truy v·∫øt ƒë∆∞·ªùng ƒëi
    if found:
        path = []
        node = end
        while node is not None:
            path.append(node)
            node = parent[node]
        path.reverse()

        print(f"\nüó∫Ô∏è ƒê∆Ø·ªúNG ƒêI T√åM ƒê∆Ø·ª¢C (DFS):")
        print(f"   {' ‚Üí '.join(map(str, path))}")
        print(f"   ƒê·ªô d√†i: {len(path)} ƒë·ªânh")

        # Visualize ƒë∆∞·ªùng ƒëi
        path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
        visualize_graph(highlight_nodes=path, highlight_edges=path_edges,
                      title=f"B√†i 2: ƒê∆∞·ªùng ƒëi {start} ‚Üí {end} (DFS)")

        return path
    else:
        print(f"üí• KH√îNG T√åM TH·∫§Y ƒê∆Ø·ªúNG ƒêI")
        return None

# CH·∫†Y B√ÄI 2
path = dfs_path_finding_demo(graph, 3, 8)

print("\n" + "="*50)
print("=== B√ÄI 3: ƒê·ªí TH·ªä PH√ÇN ƒê√îI ===")

def is_bipartite_demo(graph):
    """Minh h·ªça ki·ªÉm tra ƒë·ªì th·ªã ph√¢n ƒë√¥i b·∫±ng DFS"""
    color = {}  # 0: ch∆∞a t√¥, 1: m√†u A, 2: m√†u B
    bipartite = True

    print("üé® KI·ªÇM TRA ƒê·ªí TH·ªä PH√ÇN ƒê√îI B·∫∞NG DFS")

    for node in sorted(graph.keys()):
        if node not in color:
            print(f"\n--- B·∫Øt ƒë·∫ßu t·ª´ th√†nh ph·∫ßn li√™n th√¥ng ch·ª©a ƒë·ªânh {node} ---")

            stack = [node]
            color[node] = 1
            print(f"   T√¥ ƒë·ªânh {node} m√†u 1")

            while stack:
                current = stack.pop()
                current_color = color[current]
                next_color = 3 - current_color  # 1‚Üí2, 2‚Üí1

                print(f"\n   üîç X√©t ƒë·ªânh {current} (m√†u {current_color})")
                print(f"      C√°c ƒë·ªânh k·ªÅ: {graph[current]}")

                for neighbor in graph[current]:
                    if neighbor not in color:
                        color[neighbor] = next_color
                        stack.append(neighbor)
                        print(f"      ‚úÖ T√¥ ƒë·ªânh {neighbor} m√†u {next_color}")
                    else:
                        if color[neighbor] == current_color:
                            print(f"      üí• XUNG ƒê·ªòT: ƒê·ªânh {neighbor} ƒë√£ c√≥ m√†u {color[neighbor]}")
                            bipartite = False
                        else:
                            print(f"      üëç Ph√π h·ª£p: ƒê·ªânh {neighbor} c√≥ m√†u {color[neighbor]}")

    print(f"\nüìä K·∫æT LU·∫¨N B√ÄI 3:")
    if bipartite:
        print("   ‚úÖ ƒê·ªí TH·ªä PH√ÇN ƒê√îI")

        # Hi·ªÉn th·ªã 2 t·∫≠p
        set_A = [node for node, c in color.items() if c == 1]
        set_B = [node for node, c in color.items() if c == 2]
        print(f"   T·∫≠p A (m√†u 1): {sorted(set_A)}")
        print(f"   T·∫≠p B (m√†u 2): {sorted(set_B)}")
    else:
        print("   ‚ùå ƒê·ªí TH·ªä KH√îNG PH√ÇN ƒê√îI")

    # Visualize v·ªõi m√†u ph√¢n ƒë√¥i
    if bipartite:
        bipartite_colors = {node: 'red' if color[node] == 1 else 'blue' for node in color}
        visualize_graph(bipartite_colors=bipartite_colors,
                      title="B√†i 3: ƒê·ªì th·ªã ph√¢n ƒë√¥i\nT·∫≠p A (ƒë·ªè) - T·∫≠p B (xanh)")
    else:
        visualize_graph(title="B√†i 3: ƒê·ªì th·ªã KH√îNG ph√¢n ƒë√¥i")

    return bipartite, color

# CH·∫†Y B√ÄI 3
is_bipartite, color_map = is_bipartite_demo(graph)

print("\n" + "="*50)
print("=== B√ÄI 4: DFS KH√îNG ƒê·ªÜ QUY - SO S√ÅNH ===")

def dfs_recursive_demo(graph, node, visited, steps, depth=0):
    """Minh h·ªça DFS ƒë·ªá quy"""
    if node in visited:
        return

    visited.add(node)
    steps.append(f"{'  ' * depth}DFS({node})")
    print(f"{'  ' * depth}‚ñ∂ DFS({node})")

    for neighbor in graph[node]:
        if neighbor not in visited:
            steps.append(f"{'  ' * (depth+1)}‚Üí G·ªçi DFS({neighbor})")
            print(f"{'  ' * (depth+1)}‚Ü≥ G·ªçi DFS({neighbor})")
            dfs_recursive_demo(graph, neighbor, visited, steps, depth + 1)

    steps.append(f"{'  ' * depth}K·∫øt th√∫c DFS({node})")
    print(f"{'  ' * depth}‚óÄ K·∫øt th√∫c DFS({node})")

def dfs_non_recursive_comparison(graph, start):
    """So s√°nh DFS ƒë·ªá quy v√† kh√¥ng ƒë·ªá quy"""
    print("üîÑ SO S√ÅNH DFS ƒê·ªÜ QUY VS KH√îNG ƒê·ªÜ QUY")

    print(f"\n1. DFS ƒê·ªÜ QUY t·ª´ ƒë·ªânh {start}:")
    visited_recursive = set()
    steps_recursive = []
    dfs_recursive_demo(graph, start, visited_recursive, steps_recursive)

    print(f"\n   K·∫øt qu·∫£: {sorted(visited_recursive)}")

    print(f"\n2. DFS KH√îNG ƒê·ªÜ QUY t·ª´ ƒë·ªânh {start}:")
    visited_non_recursive = set()
    stack = [start]
    steps_non_recursive = []

    step_count = 1
    while stack:
        current = stack.pop()
        if current not in visited_non_recursive:
            visited_non_recursive.add(current)
            steps_non_recursive.append(f"B∆∞·ªõc {step_count}: L·∫•y {current} t·ª´ stack")
            print(f"   B∆∞·ªõc {step_count}: L·∫•y {current} t·ª´ stack")
            step_count += 1

            # Th√™m c√°c ƒë·ªânh k·ªÅ theo th·ª© t·ª± ng∆∞·ª£c
            for neighbor in reversed(graph[current]):
                if neighbor not in visited_non_recursive:
                    stack.append(neighbor)
                    print(f"      ‚Üí Th√™m {neighbor} v√†o stack")

    print(f"\n   K·∫øt qu·∫£: {sorted(visited_non_recursive)}")

    print(f"\nüìä SO S√ÅNH K·∫æT QU·∫¢:")
    print(f"   DFS ƒë·ªá quy:    {sorted(visited_recursive)}")
    print(f"   DFS kh√¥ng ƒë·ªá quy: {sorted(visited_non_recursive)}")
    print(f"   K·∫øt qu·∫£ {'GI·ªêNG' if visited_recursive == visited_non_recursive else 'KH√ÅC'} nhau")

    return visited_recursive, visited_non_recursive

# CH·∫†Y B√ÄI 4
visited_rec, visited_non_rec = dfs_non_recursive_comparison(graph, 7)

print("\n" + "="*60)
print("üéä T·ªîNG K·∫æT 4 B√ÄI T·∫¨P DFS")
print("="*60)

print(f"\nüìã B√ÄI 1 - Li√™n th√¥ng v·ªõi ƒë·ªânh 7 (DFS):")
print(f"   K·∫øt qu·∫£: {sorted(visited_nodes)}")

print(f"\nüìã B√ÄI 2 - ƒê∆∞·ªùng ƒëi t·ª´ 3 ƒë·∫øn 8 (DFS):")
print(f"   K·∫øt qu·∫£: {' ‚Üí '.join(map(str, path))}")

print(f"\nüìã B√ÄI 3 - ƒê·ªì th·ªã ph√¢n ƒë√¥i:")
print(f"   K·∫øt qu·∫£: {'PH√ÇN ƒê√îI' if is_bipartite else 'KH√îNG PH√ÇN ƒê√îI'}")
if is_bipartite:
    set_A = [node for node, c in color_map.items() if c == 1]
    set_B = [node for node, c in color_map.items() if c == 2]
    print(f"   T·∫≠p A: {sorted(set_A)}")
    print(f"   T·∫≠p B: {sorted(set_B)}")

print(f"\nüìã B√ÄI 4 - So s√°nh DFS:")
print(f"   DFS ƒë·ªá quy v√† kh√¥ng ƒë·ªá quy: {'GI·ªêNG' if visited_rec == visited_non_rec else 'KH√ÅC'}")

print(f"\nüéØ ƒê·∫∂C ƒêI·ªÇM DFS:")
print(f"   - D√πng stack (ƒë·ªá quy ho·∫∑c kh√¥ng ƒë·ªá quy)")
print(f"   - ∆Øu ti√™n ƒëi s√¢u theo 1 nh√°nh")
print(f"   - T√¨m ƒë∆∞·ªùng ƒëi c√≥ th·ªÉ kh√¥ng ng·∫Øn nh·∫•t")
print(f"   - Ph√π h·ª£p cho ki·ªÉm tra chu tr√¨nh, ƒë·ªì th·ªã ph√¢n ƒë√¥i")

print("\n" + "="*60)
print("‚ú® K·∫æT TH√öC CH∆Ø∆†NG TR√åNH MINH H·ªåA DFS ‚ú®")
print("="*60)
```
