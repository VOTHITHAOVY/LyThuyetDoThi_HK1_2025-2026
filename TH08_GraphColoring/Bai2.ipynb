````python
# %% [markdown]
# # B√ÄI 2: T√î M√ÄU B·∫¢N ƒê·ªí C√ÅC BANG N∆Ø·ªöC M·ª∏

# %% [markdown]
# ## üìå √ù T∆Ø·ªûNG CH√çNH

# %% [markdown]
# ### **B√†i To√°n T√¥ M√†u B·∫£n ƒê·ªì**
# **M·ª•c ti√™u**: T√¥ m√†u c√°c bang sao cho:
# - Hai bang k·ªÅ nhau kh√¥ng c√πng m√†u
# - S·ªë m√†u s·ª≠ d·ª•ng l√† √≠t nh·∫•t
#
# ### üéØ **Thu·∫≠t To√°n Tham Lam**
# ```
# 1. S·∫Øp x·∫øp c√°c bang theo s·ªë bang k·ªÅ gi·∫£m d·∫ßn
# 2. T√¥ m√†u bang c√≥ nhi·ªÅu bang k·ªÅ tr∆∞·ªõc
# 3. Ch·ªçn m√†u nh·ªè nh·∫•t c√≥ th·ªÉ cho m·ªói bang
# ```
#
# ### üó∫Ô∏è **C√°c Bang C·∫ßn T√¥ M√†u**
# - NEVADA, UTAH, COLORADO, CALIFORNIA, ARIZONA
# - NEBRASKA, KANSAS, OKLAHOMA, ARKANSAS, MISSISSIPI
# - GEORGIA, TEXAS, LOUISIANA, FLORIDA

# %% [markdown]
# ## üöÄ MINH H·ªåA CH·∫†Y

# %%
# KH·ªûI T·∫†O D·ªÆ LI·ªÜU
print("=" * 60)
print("KH·ªûI T·∫†O D·ªÆ LI·ªÜU B·∫¢N ƒê·ªí")
print("=" * 60)

states = [
    "NEVADA", "UTAH", "COLORADO", "CALIFORNIA", "ARIZONA",
    "NEBRASKA", "KANSAS", "OKLAHOMA", "ARKANSAS", "MISSISSIPI",
    "GEORGIA", "TEXAS", "LOUISIANA", "FLORIDA"
]

print("Danh s√°ch c√°c bang:")
for i, state in enumerate(states, 1):
    print(f"{i:2d}. {state}")

# %%
# X√ÇY D·ª∞NG ƒê·ªí TH·ªä QUAN H·ªÜ K·ªÄ NHAU
print("\n" + "=" * 60)
print("QUAN H·ªÜ K·ªÄ NHAU GI·ªÆA C√ÅC BANG")
print("=" * 60)

# T·∫°o dictionary l∆∞u quan h·ªá k·ªÅ nhau
borders = {
    "NEVADA": ["UTAH", "CALIFORNIA", "ARIZONA"],
    "UTAH": ["NEVADA", "COLORADO", "ARIZONA"],
    "COLORADO": ["UTAH", "ARIZONA", "NEBRASKA", "KANSAS", "OKLAHOMA"],
    "CALIFORNIA": ["NEVADA", "ARIZONA"],
    "ARIZONA": ["NEVADA", "UTAH", "CALIFORNIA", "COLORADO"],
    "NEBRASKA": ["COLORADO", "KANSAS"],
    "KANSAS": ["NEBRASKA", "COLORADO", "OKLAHOMA"],
    "OKLAHOMA": ["COLORADO", "KANSAS", "TEXAS", "ARKANSAS"],
    "ARKANSAS": ["OKLAHOMA", "TEXAS", "LOUISIANA", "MISSISSIPI"],
    "MISSISSIPI": ["ARKANSAS", "LOUISIANA", "GEORGIA"],
    "GEORGIA": ["MISSISSIPI", "FLORIDA"],
    "TEXAS": ["OKLAHOMA", "ARKANSAS", "LOUISIANA"],
    "LOUISIANA": ["TEXAS", "ARKANSAS", "MISSISSIPI"],
    "FLORIDA": ["GEORGIA"]
}

# Hi·ªÉn th·ªã quan h·ªá k·ªÅ nhau
for state, neighbors in borders.items():
    print(f"{state:12} k·ªÅ v·ªõi: {', '.join(neighbors)}")

# %%
# T√çNH B·∫¨C C√ÅC BANG (S·ªê BANG K·ªÄ)
print("\n" + "=" * 60)
print("T√çNH B·∫¨C C√ÅC BANG")
print("=" * 60)

degrees = {}
for state in states:
    degrees[state] = len(borders[state])

# S·∫Øp x·∫øp theo b·∫≠c gi·∫£m d·∫ßn
sorted_degrees = sorted(degrees.items(), key=lambda x: x[1], reverse=True)

print("B·∫≠c c√°c bang (s·ªë bang k·ªÅ):")
for state, degree in sorted_degrees:
    print(f"{state:12}: {degree:2} bang k·ªÅ")

# %%
# THU·∫¨T TO√ÅN T√î M√ÄU THAM LAM
print("\n" + "=" * 60)
print("THU·∫¨T TO√ÅN T√î M√ÄU THAM LAM")
print("=" * 60)

colors = {}  # L∆∞u m√†u c·ªßa t·ª´ng bang
color_names = {1: "üî¥ ƒê·ªé", 2: "üîµ XANH", 3: "üü¢ L·ª§C", 4: "üü° V√ÄNG", 5: "üü£ T√çM"}

# T√¥ m√†u theo th·ª© t·ª± ƒë√£ s·∫Øp x·∫øp
for state, _ in sorted_degrees:
    used_colors = set()

    # T√¨m m√†u ƒë√£ d√πng b·ªüi c√°c bang k·ªÅ
    for neighbor in borders[state]:
        if neighbor in colors:
            used_colors.add(colors[neighbor])

    # T√¨m m√†u nh·ªè nh·∫•t ch∆∞a d√πng
    current_color = 1
    while current_color in used_colors:
        current_color += 1

    colors[state] = current_color

    # Hi·ªÉn th·ªã t·ª´ng b∆∞·ªõc
    used_str = ", ".join(map(str, sorted(used_colors))) if used_colors else "kh√¥ng c√≥"
    print(f"{state:12} ‚Üí M√†u {current_color} (m√†u c·∫•m: {used_str})")

# %%
# K·∫æT QU·∫¢ CU·ªêI C√ôNG
print("\n" + "=" * 60)
print("K·∫æT QU·∫¢ T√î M√ÄU")
print("=" * 60)

# Nh√≥m c√°c bang theo m√†u
color_groups = {}
for state, color in colors.items():
    if color not in color_groups:
        color_groups[color] = []
    color_groups[color].append(state)

# Hi·ªÉn th·ªã k·∫øt qu·∫£
num_colors = len(color_groups)
print(f"üé® S·ªë m√†u s·ª≠ d·ª•ng: {num_colors}")

for color_num in sorted(color_groups.keys()):
    states_in_color = sorted(color_groups[color_num])
    print(f"\n{color_names.get(color_num, f'M√†u {color_num}')}:")
    for state in states_in_color:
        print(f"  ‚Ä¢ {state}")

# %%
# KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN
print("\n" + "=" * 60)
print("KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN")
print("=" * 60)

valid = True
error_pairs = []

for state, neighbors in borders.items():
    for neighbor in neighbors:
        if colors[state] == colors[neighbor]:
            valid = False
            error_pairs.append((state, neighbor, colors[state]))

if valid:
    print("‚úÖ T√¥ m√†u H·ª¢P L·ªÜ: Kh√¥ng c√≥ hai bang k·ªÅ nhau c√πng m√†u")
else:
    print("‚ùå T√¥ m√†u KH√îNG H·ª¢P L·ªÜ:")
    for state1, state2, color in error_pairs:
        print(f"  - {state1} v√† {state2} c√πng m√†u {color}")

# %%
# TR·ª∞C QUAN H√ìA K·∫æT QU·∫¢
print("\n" + "=" * 60)
print("TR·ª∞C QUAN H√ìA B·∫¢N ƒê·ªí")
print("=" * 60)

# T·∫°o bi·ªÉu di·ªÖn tr·ª±c quan ƒë∆°n gi·∫£n
print("\nB·∫¢N ƒê·ªí ƒê√É T√î M√ÄU:")
print("(M·ªói bang ƒë∆∞·ª£c bi·ªÉu di·ªÖn b·∫±ng s·ªë m√†u t∆∞∆°ng ·ª©ng)")

# T·∫°o grid ƒë∆°n gi·∫£n ƒë·ªÉ hi·ªÉn th·ªã
visual_map = [
    ["NEBRASKA", "KANSAS", "OKLAHOMA", "ARKANSAS", "MISSISSIPI", "GEORGIA"],
    ["", "COLORADO", "", "TEXAS", "LOUISIANA", "FLORIDA"],
    ["UTAH", "", "ARIZONA", "", "", ""],
    ["NEVADA", "CALIFORNIA", "", "", "", ""]
]

print("\n" + "-" * 70)
for row in visual_map:
    for state in row:
        if state:
            color_num = colors[state]
            color_symbol = "üî¥üîµüü¢üü°üü£"[color_num-1]
            print(f"{color_symbol}{color_num}-{state[:3]:3}", end="  ")
        else:
            print(" " * 10, end="  ")
    print()
print("-" * 70)

# %%
# CH√ö TH√çCH M√ÄU
print("\nCH√ö TH√çCH M√ÄU:")
for color_num, color_name in color_names.items():
    if color_num <= num_colors:
        print(f"{color_name} ({color_num})")

# %% [markdown]
# ## üìä K·∫æT LU·∫¨N

# %% [markdown]
# - **S·ªë m√†u s·ª≠ d·ª•ng**: 4 m√†u
# - **K·∫øt qu·∫£**: H·ª£p l·ªá - kh√¥ng c√≥ hai bang k·ªÅ nhau c√πng m√†u
# - **Thu·∫≠t to√°n tham lam** cho k·∫øt qu·∫£ t·ªët v·ªõi b√†i to√°n t√¥ m√†u b·∫£n ƒë·ªì
# - **·ª®ng d·ª•ng**: C√≥ th·ªÉ √°p d·ª•ng cho c√°c b√†i to√°n l·∫≠p l·ªãch, ph√¢n b·ªï t√†i nguy√™n

# %%
# L∆ØU K·∫æT QU·∫¢
print("\n" + "=" * 60)
print("L∆ØU K·∫æT QU·∫¢")
print("=" * 60)

# T·∫°o file k·∫øt qu·∫£
result_content = f"{num_colors}\n"
for color_num in sorted(color_groups.keys()):
    states_line = " ".join(sorted(color_groups[color_num]))
    result_content += f"{states_line}\n"

with open('ketqua_tomaubando.txt', 'w', encoding='utf-8') as f:
    f.write(result_content)

print("‚úÖ ƒê√£ l∆∞u k·∫øt qu·∫£ v√†o file 'ketqua_tomaubando.txt'")
print("\nN·ªôi dung file:")
print(result_content)
````
