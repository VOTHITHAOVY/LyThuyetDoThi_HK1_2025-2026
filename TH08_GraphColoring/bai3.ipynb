````python
# %% [markdown]
# # B√ÄI 3: X√ÇY D·ª∞NG ƒê·ªòI NG≈® (BUILDING_TEAMS)

# %% [markdown]
# ## üìå √ù T∆Ø·ªûNG CH√çNH

# %% [markdown]
# ### **B√†i To√°n Chia ƒê·ªôi**
# **M·ª•c ti√™u**: Chia n h·ªçc sinh th√†nh 2 ƒë·ªôi sao cho:
# - Kh√¥ng c√≥ hai b·∫°n b√® n√†o c√πng ƒë·ªôi
# - N·∫øu kh√¥ng th·ªÉ chia, in "IMPOSSIBLE"
#
# ### üéØ **Thu·∫≠t To√°n Ki·ªÉm Tra ƒê·ªì Th·ªã Hai Ph√≠a**
# ```
# 1. M·ªói h·ªçc sinh = 1 ƒë·ªânh ƒë·ªì th·ªã
# 2. Quan h·ªá b·∫°n b√® = 1 c·∫°nh n·ªëi
# 3. Chia 2 ƒë·ªôi = T√¥ 2 m√†u cho ƒë·ªì th·ªã
# 4. ƒêi·ªÅu ki·ªán: Kh√¥ng c√≥ c·∫°nh n·ªëi 2 ƒë·ªânh c√πng m√†u
# ```
#
# ### ‚ö° **ƒê·ªô Ph·ª©c T·∫°p**
# - **Th·ªùi gian**: O(n + m)
# - **Kh√¥ng gian**: O(n + m)

# %% [markdown]
# ## üöÄ MINH H·ªåA CH·∫†Y

# %%
# KH·ªûI T·∫†O D·ªÆ LI·ªÜU
print("=" * 60)
print("KH·ªûI T·∫†O D·ªÆ LI·ªÜU")
print("=" * 60)

# D·ªØ li·ªáu t·ª´ v√≠ d·ª•
n = 5
m = 3
friendships = [(1, 2), (1, 3), (4, 5)]

print(f"S·ªë h·ªçc sinh: {n}")
print(f"S·ªë quan h·ªá b·∫°n b√®: {m}")
print("Danh s√°ch b·∫°n b√®:")
for i, (a, b) in enumerate(friendships, 1):
    print(f"  {i}. {a} - {b}")

# %%
# X√ÇY D·ª∞NG ƒê·ªí TH·ªä
print("\n" + "=" * 60)
print("X√ÇY D·ª∞NG ƒê·ªí TH·ªä")
print("=" * 60)

# T·∫°o danh s√°ch k·ªÅ
adj = [[] for _ in range(n + 1)]

for a, b in friendships:
    adj[a].append(b)
    adj[b].append(a)

print("Danh s√°ch k·ªÅ:")
for i in range(1, n + 1):
    print(f"H·ªçc sinh {i}: k·ªÅ v·ªõi {adj[i]}")

# %%
# THU·∫¨T TO√ÅN BFS KI·ªÇM TRA ƒê·ªí TH·ªä HAI PH√çA
print("\n" + "=" * 60)
print("THU·∫¨T TO√ÅN BFS KI·ªÇM TRA ƒê·ªí TH·ªä HAI PH√çA")
print("=" * 60)

from collections import deque

color = [0] * (n + 1)  # 0: ch∆∞a t√¥, 1: ƒë·ªôi 1, 2: ƒë·ªôi 2
is_possible = True

def bfs(start):
    queue = deque([start])
    color[start] = 1  # B·∫Øt ƒë·∫ßu v·ªõi ƒë·ªôi 1

    while queue:
        u = queue.popleft()
        current_team = color[u]
        next_team = 3 - current_team  # ƒê·ªôi ƒë·ªëi l·∫≠p

        print(f"\nX√©t h·ªçc sinh {u} (ƒë·ªôi {current_team}):")

        for v in adj[u]:
            if color[v] == 0:
                # Ch∆∞a ph√¢n ƒë·ªôi ‚Üí ph√¢n ƒë·ªôi ƒë·ªëi l·∫≠p
                color[v] = next_team
                queue.append(v)
                print(f"  ‚Üí H·ªçc sinh {v} ch∆∞a c√≥ ƒë·ªôi ‚Üí ph√¢n ƒë·ªôi {next_team}")
            elif color[v] == current_team:
                # Xung ƒë·ªôt: b·∫°n b√® c√πng ƒë·ªôi
                print(f"  ‚ùå XUNG ƒê·ªòT: {u} v√† {v} c√πng ƒë·ªôi {current_team}")
                return False
            else:
                print(f"  ‚Üí H·ªçc sinh {v} ƒë√£ ·ªü ƒë·ªôi {color[v]} (h·ª£p l·ªá)")

    return True

# Ch·∫°y BFS cho t·∫•t c·∫£ th√†nh ph·∫ßn li√™n th√¥ng
for i in range(1, n + 1):
    if color[i] == 0:  # Ch∆∞a ƒë∆∞·ª£c thƒÉm
        print(f"\n--- B·∫Øt ƒë·∫ßu t·ª´ h·ªçc sinh {i} ---")
        if not bfs(i):
            is_possible = False
            break

# %%
# K·∫æT QU·∫¢
print("\n" + "=" * 60)
print("K·∫æT QU·∫¢")
print("=" * 60)

if not is_possible:
    print("‚ùå IMPOSSIBLE")
    print("Kh√¥ng th·ªÉ chia ƒë·ªôi v√¨ c√≥ hai b·∫°n b√® ph·∫£i c√πng ƒë·ªôi!")
else:
    print("‚úÖ C√ì TH·ªÇ CHIA ƒê·ªòI")
    print("\nPh√¢n b·ªï ƒë·ªôi:")
    team1 = []
    team2 = []

    for i in range(1, n + 1):
        if color[i] == 1:
            team1.append(i)
        else:
            team2.append(i)

    print(f"ƒê·ªôi 1: {team1}")
    print(f"ƒê·ªôi 2: {team2}")

    print("\nOutput theo y√™u c·∫ßu:")
    output = [str(color[i]) for i in range(1, n + 1)]
    print(" ".join(output))

# %%
# KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN
print("\n" + "=" * 60)
print("KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN")
print("=" * 60)

if is_possible:
    valid = True
    print("Ki·ªÉm tra t·ª´ng quan h·ªá b·∫°n b√®:")

    for a, b in friendships:
        if color[a] == color[b]:
            print(f"‚ùå L·ªói: {a} v√† {b} c√πng ƒë·ªôi {color[a]}")
            valid = False
        else:
            print(f"‚úÖ {a} (ƒë·ªôi {color[a]}) v√† {b} (ƒë·ªôi {color[b]}) - H·ª£p l·ªá")

    if valid:
        print("\nüéâ T·∫•t c·∫£ ƒë·ªÅu h·ª£p l·ªá! Kh√¥ng c√≥ b·∫°n b√® n√†o c√πng ƒë·ªôi.")
else:
    print("Kh√¥ng c·∫ßn ki·ªÉm tra v√¨ kh√¥ng th·ªÉ chia ƒë·ªôi")

# %%
# TEST CASE 2: TR∆Ø·ªúNG H·ª¢P IMPOSSIBLE
print("\n" + "=" * 60)
print("TEST CASE 2: TR∆Ø·ªúNG H·ª¢P IMPOSSIBLE")
print("=" * 60)

# D·ªØ li·ªáu c√≥ chu tr√¨nh l·∫ª
n2 = 4
m2 = 4
friendships2 = [(1, 2), (1, 3), (2, 3), (3, 4)]

print("ƒê·ªì th·ªã c√≥ chu tr√¨nh tam gi√°c (1-2-3)")
print("Danh s√°ch b·∫°n b√®:", friendships2)

# Reset
adj2 = [[] for _ in range(n2 + 1)]
color2 = [0] * (n2 + 1)

for a, b in friendships2:
    adj2[a].append(b)
    adj2[b].append(a)

# Ch·∫°y BFS
is_possible2 = True
for i in range(1, n2 + 1):
    if color2[i] == 0:
        if not bfs(i):  # S·ª≠ d·ª•ng h√†m BFS ƒë√£ ƒë·ªãnh nghƒ©a
            is_possible2 = False
            break

print(f"\nK·∫øt qu·∫£: {'IMPOSSIBLE' if not is_possible2 else 'C√ì TH·ªÇ CHIA'}")

# %%
# TR·ª∞C QUAN H√ìA ƒê·ªí TH·ªä
print("\n" + "=" * 60)
print("TR·ª∞C QUAN H√ìA ƒê·ªí TH·ªä")
print("=" * 60)

try:
    import networkx as nx
    import matplotlib.pyplot as plt

    # T·∫°o ƒë·ªì th·ªã
    G = nx.Graph()

    # Th√™m c√°c c·∫°nh
    for a, b in friendships:
        G.add_edge(a, b)

    # T·∫°o m√†u cho nodes
    node_colors = []
    for i in range(1, n + 1):
        if color[i] == 1:
            node_colors.append('red')
        else:
            node_colors.append('blue')

    # V·∫Ω ƒë·ªì th·ªã
    plt.figure(figsize=(10, 6))
    pos = nx.spring_layout(G)

    nx.draw(G, pos, with_labels=True, node_color=node_colors,
            node_size=800, font_size=12, font_weight='bold',
            edge_color='gray', width=2)

    plt.title("ƒê·ªí TH·ªä QUAN H·ªÜ B·∫†N B√à V√Ä PH√ÇN ƒê·ªòI", fontsize=14)
    plt.figtext(0.5, 0.01, "üî¥ ƒê·ªôi 1 | üîµ ƒê·ªôi 2", ha="center", fontsize=12)
    plt.show()

except ImportError:
    print("‚ö†Ô∏è C·∫ßn c√†i ƒë·∫∑t: pip install networkx matplotlib")

# %%
# SO S√ÅNH C√ÅC PH∆Ø∆†NG PH√ÅP
print("\n" + "=" * 60)
print("SO S√ÅNH C√ÅC PH∆Ø∆†NG PH√ÅP")
print("=" * 60)

print("""
PH∆Ø∆†NG PH√ÅP BFS:
- ∆Øu ƒëi·ªÉm: ƒê∆°n gi·∫£n, d·ªÖ c√†i ƒë·∫∑t
- Hi·ªáu qu·∫£: O(n + m) - ph√π h·ª£p v·ªõi d·ªØ li·ªáu l·ªõn
- ·ª®ng d·ª•ng: Ki·ªÉm tra ƒë·ªì th·ªã hai ph√≠a

PH∆Ø∆†NG PH√ÅP DFS:
- T∆∞∆°ng t·ª± BFS, c√≥ th·ªÉ s·ª≠ d·ª•ng ƒë·ªá quy
- C√πng ƒë·ªô ph·ª©c t·∫°p O(n + m)

PH∆Ø∆†NG PH√ÅP DSU (Disjoint Set Union):
- C√≥ th·ªÉ s·ª≠ d·ª•ng nh∆∞ng ph·ª©c t·∫°p h∆°n
- Kh√¥ng tr·ª±c quan b·∫±ng BFS/DFS
""")

# %%
# X·ª¨ L√ù INPUT L·ªöN
print("\n" + "=" * 60)
print("X·ª¨ L√ù INPUT L·ªöN")
print("=" * 60)

print(f"Gi·ªõi h·∫°n ƒë·ªÅ b√†i:")
print(f"- n ‚â§ {10**5:,} h·ªçc sinh")
print(f"- m ‚â§ {2 * 10**5:,} quan h·ªá b·∫°n b√®")

print(f"\nƒê·ªô ph·ª©c t·∫°p thu·∫≠t to√°n:")
print(f"- O(n + m) = O({10**5 + 2*10**5:,})")
print(f"- ƒê·ªß nhanh cho gi·ªõi h·∫°n ƒë·ªÅ b√†i")

# %%
# L∆ØU K·∫æT QU·∫¢
print("\n" + "=" * 60)
print("L∆ØU K·∫æT QU·∫¢")
print("=" * 60)

if is_possible:
    result = " ".join(str(color[i]) for i in range(1, n + 1))
    with open('ketqua_building_teams.txt', 'w') as f:
        f.write(result)
    print("‚úÖ ƒê√£ l∆∞u k·∫øt qu·∫£ v√†o 'ketqua_building_teams.txt'")
    print("N·ªôi dung:", result)
else:
    with open('ketqua_building_teams.txt', 'w') as f:
        f.write("IMPOSSIBLE")
    print("‚úÖ ƒê√£ l∆∞u k·∫øt qu·∫£: IMPOSSIBLE")

# %% [markdown]
# ## üìä K·∫æT LU·∫¨N

# %% [markdown]
# - **B√†i to√°n** t∆∞∆°ng ƒë∆∞∆°ng v·ªõi **ki·ªÉm tra ƒë·ªì th·ªã hai ph√≠a**
# - **Thu·∫≠t to√°n BFS** hi·ªáu qu·∫£ v·ªõi ƒë·ªô ph·ª©c t·∫°p O(n + m)
# - **·ª®ng d·ª•ng** r·ªông r√£i trong chia ƒë·ªôi, x·∫øp l·ªãch, ph√¢n c√¥ng c√¥ng vi·ªác
# - **K·∫øt qu·∫£**: C√≥ th·ªÉ chia ƒë·ªôi ho·∫∑c IMPOSSIBLE t√πy v√†o t√≠nh ch·∫•t ƒë·ªì th·ªã
````
