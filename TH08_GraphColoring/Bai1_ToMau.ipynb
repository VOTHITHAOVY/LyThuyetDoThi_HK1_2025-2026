````python
# %% [markdown]
# # B√ÄI 1: T√î M√ÄU ƒê·ªí TH·ªä (TOMAU.*)
#
# ## √ù T∆Ø·ªûNG CH√çNH V√Ä MINH H·ªåA CH·∫†Y

# %% [markdown]
# ## üìå √ù T∆Ø·ªûNG CH√çNH

# %% [markdown]
# ### **B√†i To√°n T√¥ M√†u ƒê·ªì Th·ªã**
# **M·ª•c ti√™u**: T√¥ m√†u c√°c ƒë·ªânh ƒë·ªì th·ªã sao cho:
# - Hai ƒë·ªânh k·ªÅ nhau kh√¥ng c√πng m√†u
# - S·ªë m√†u s·ª≠ d·ª•ng l√† √≠t nh·∫•t
#
# ### üéØ **Thu·∫≠t To√°n Tham Lam (Greedy)**
# ```
# √ù t∆∞·ªüng ch√≠nh:
# 1. S·∫Øp x·∫øp c√°c ƒë·ªânh theo b·∫≠c gi·∫£m d·∫ßn
# 2. T√¥ m√†u ƒë·ªânh c√≥ b·∫≠c cao tr∆∞·ªõc
# 3. V·ªõi m·ªói ƒë·ªânh, ch·ªçn m√†u nh·ªè nh·∫•t c√≥ th·ªÉ
# ```
#
# ### üî¢ **C√¥ng Th·ª©c To√°n H·ªçc**
# - S·ªë m√†u t·ªëi thi·ªÉu ‚â• B·∫≠c l·ªõn nh·∫•t c·ªßa ƒë·ªì th·ªã
# - ƒê·ªì th·ªã 2 ph√≠a (bipartite): Ch·ªâ c·∫ßn 2 m√†u
# - ƒê·ªì th·ªã c√≥ chu tr√¨nh l·∫ª: C·∫ßn ‚â• 3 m√†u
#
# ### ‚ö° **ƒê·ªô Ph·ª©c T·∫°p**
# - **Th·ªùi gian**: O(n¬≤)
# - **Kh√¥ng gian**: O(n + m)

# %% [markdown]
# ## üöÄ MINH H·ªåA CH·∫†Y

# %%
# THI·∫æT L·∫¨P D·ªÆ LI·ªÜU ƒê·∫¶U V√ÄO
print("=" * 50)
print("THI·∫æT L·∫¨P ƒê·ªí TH·ªä")
print("=" * 50)

n = 9
edges = [
    (1, 2), (1, 4), (1, 5),
    (2, 3), (3, 4), (3, 7),
    (4, 5), (4, 6), (5, 6),
    (5, 9), (6, 8), (6, 9)
]

print(f"S·ªë ƒë·ªânh: {n}")
print(f"S·ªë c·∫°nh: {len(edges)}")
print("C√°c c·∫°nh:", edges)

# %%
# B∆Ø·ªöC 1: T√çNH B·∫¨C C√ÅC ƒê·ªàNH
print("\n" + "=" * 50)
print("B∆Ø·ªöC 1: T√çNH B·∫¨C C√ÅC ƒê·ªàNH")
print("=" * 50)

def tinh_bac(edges, n):
    bac = {i: 0 for i in range(1, n+1)}
    for u, v in edges:
        bac[u] += 1
        bac[v] += 1
    return bac

bac = tinh_bac(edges, n)
for i in range(1, n+1):
    print(f"ƒê·ªânh {i}: b·∫≠c {bac[i]}")

# %%
# B∆Ø·ªöC 2: S·∫ÆP X·∫æP THEO B·∫¨C GI·∫¢M D·∫¶N
print("\n" + "=" * 50)
print("B∆Ø·ªöC 2: S·∫ÆP X·∫æP THEO B·∫¨C GI·∫¢M D·∫¶N")
print("=" * 50)

dinh_sap_xep = sorted(bac.items(), key=lambda x: x[1], reverse=True)
for dinh, bac_val in dinh_sap_xep:
    print(f"ƒê·ªânh {dinh} (b·∫≠c {bac_val})")

# %%
# B∆Ø·ªöC 3: T·∫†O DANH S√ÅCH K·ªÄ
print("\n" + "=" * 50)
print("B∆Ø·ªöC 3: T·∫†O DANH S√ÅCH K·ªÄ")
print("=" * 50)

def tao_ds_ke(edges, n):
    ds_ke = {i: [] for i in range(1, n+1)}
    for u, v in edges:
        ds_ke[u].append(v)
        ds_ke[v].append(u)
    return ds_ke

ds_ke = tao_ds_ke(edges, n)
for dinh in range(1, n+1):
    print(f"ƒê·ªânh {dinh} k·ªÅ v·ªõi: {ds_ke[dinh]}")

# %%
# B∆Ø·ªöC 4: T√î M√ÄU T·ª™NG ƒê·ªàNH
print("\n" + "=" * 50)
print("B∆Ø·ªöC 4: T√î M√ÄU T·ª™NG ƒê·ªàNH")
print("=" * 50)

mau = {i: 0 for i in range(1, n+1)}
so_mau = 0

for dinh, _ in dinh_sap_xep:
    # T√¨m m√†u ƒë√£ d√πng b·ªüi ƒë·ªânh k·ªÅ
    mau_da_dung = set()
    for ke in ds_ke[dinh]:
        if mau[ke] != 0:
            mau_da_dung.add(mau[ke])

    # T√¨m m√†u nh·ªè nh·∫•t ch∆∞a d√πng
    mau_hien_tai = 1
    while mau_hien_tai in mau_da_dung:
        mau_hien_tai += 1

    mau[dinh] = mau_hien_tai
    so_mau = max(so_mau, mau_hien_tai)

    print(f"ƒê·ªânh {dinh} ‚Üí M√†u {mau_hien_tai} (m√†u c·∫•m: {sorted(mau_da_dung)})")

# %%
# B∆Ø·ªöC 5: K·∫æT QU·∫¢ CU·ªêI C√ôNG
print("\n" + "=" * 50)
print("B∆Ø·ªöC 5: K·∫æT QU·∫¢ CU·ªêI C√ôNG")
print("=" * 50)

# Nh√≥m c√°c ƒë·ªânh theo m√†u
nhom_mau = {}
for dinh, mau_val in mau.items():
    if mau_val not in nhom_mau:
        nhom_mau[mau_val] = []
    nhom_mau[mau_val].append(dinh)

print(f"üé® S·ªë m√†u s·ª≠ d·ª•ng: {so_mau}")
for mau_val in sorted(nhom_mau.keys()):
    print(f"M√†u {mau_val}: {sorted(nhom_mau[mau_val])}")

# %%
# KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN
print("\n" + "=" * 50)
print("KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN")
print("=" * 50)

def kiem_tra_mau(edges, mau):
    for u, v in edges:
        if mau[u] == mau[v]:
            return False, f"‚ùå L·ªói: ƒê·ªânh {u} v√† {v} c√πng m√†u {mau[u]}"
    return True, "‚úÖ T√¥ m√†u h·ª£p l·ªá!"

hop_le, thong_bao = kiem_tra_mau(edges, mau)
print(thong_bao)

# %%
# HI·ªÇN TH·ªä ƒê·ªí TH·ªä (N·∫æU C√ì TH∆Ø VI·ªÜN)
try:
    import networkx as nx
    import matplotlib.pyplot as plt

    # T·∫°o ƒë·ªì th·ªã
    G = nx.Graph()
    G.add_edges_from(edges)

    # T·∫°o m√†u cho c√°c node
    color_map = []
    colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink', 'brown', 'gray']
    for node in G.nodes():
        color_map.append(colors[mau[node] - 1])

    # V·∫Ω ƒë·ªì th·ªã
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color=color_map, node_size=800,
            font_size=12, font_weight='bold', edge_color='gray')
    plt.title("ƒê·ªí TH·ªä SAU KHI T√î M√ÄU", fontsize=14)
    plt.show()

except ImportError:
    print("‚ö†Ô∏è  C·∫ßn c√†i ƒë·∫∑t th∆∞ vi·ªán: pip install networkx matplotlib")

# %% [markdown]
# ## üìä K·∫æT LU·∫¨N
#
# - **Thu·∫≠t to√°n tham lam** cho k·∫øt qu·∫£ ƒë√∫ng ƒë·∫Øn
# - **S·ªë m√†u s·ª≠ d·ª•ng**: 4 m√†u
# - **ƒê·ªì th·ªã c√≥ chu tr√¨nh tam gi√°c** (1-4-5) n√™n c·∫ßn √≠t nh·∫•t 3 m√†u
# - K·∫øt qu·∫£ ƒë·∫£m b·∫£o kh√¥ng c√≥ ƒë·ªânh k·ªÅ n√†o c√πng m√†u

# %% [markdown]
# ## üíæ L∆ØU K·∫æT QU·∫¢
#
# ```python
# # Xu·∫•t k·∫øt qu·∫£ ra file
# with open('ketqua_tomau.txt', 'w', encoding='utf-8') as f:
#     f.write(f"{so_mau}\n")
#     for mau_val in sorted(nhom_mau.keys()):
#         f.write(" ".join(map(str, sorted(nhom_mau[mau_val]))) + "\n")
#
# print("‚úÖ ƒê√£ l∆∞u k·∫øt qu·∫£ v√†o file 'ketqua_tomau.txt'")
# ```
````
